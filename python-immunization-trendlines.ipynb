```python
import pandas as pd
filename = '/kaggle/input/school-based-immunization-coverage-in-nova-scotia/School-Based_Immunization_Coverage_in_Nova_Scotia.csv'
df = pd.read_csv(filepath_or_buffer=filename, thousands=',')
df['year']= df['Year'].apply(func=lambda x: int(x.split('-')[0]))
df.head()
```

```python
from plotly.express import bar
bar(data_frame=df, x='Year', color='Vaccine')
```

```python
bar(data_frame=df, x='Year', color='Zone')
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='year', y='% Coverage', color='Vaccine', symbol='Zone')
```

```python
grouped_df = df[['year', 'Vaccine', '# Immunized', '# Eligible']].groupby(by=['year', 'Vaccine']).sum().reset_index()
grouped_df['pct_immunized'] = grouped_df['# Immunized'] / grouped_df['# Eligible']
```

```python
scatter(data_frame=grouped_df, x='# Eligible', y='# Immunized', color='Vaccine', symbol='year')
```

```python
scatter(data_frame=grouped_df, y='pct_immunized', color='Vaccine', x='year', trendline='lowess')
```

This is probably the nut graf here as it shows immunization rates declining.
