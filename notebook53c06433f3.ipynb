```python
import pandas as pd
CUSTOMER = '/kaggle/input/customer-segmentation-clustering/customer_segmentation.csv'
df = pd.read_csv(filepath_or_buffer=CUSTOMER, index_col=['ID'], parse_dates=['Dt_Customer'])
df['date'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
df = df.drop(columns=['Dt_Customer', 'Z_CostContact', 'Z_Revenue'])
df.head()
```

```python
df.info()
```

```python
from plotly.express import histogram
for column in df.columns:
    histogram(data_frame=df, x=column, color='Response', log_y=column.startswith('Mnt')).show()
```
