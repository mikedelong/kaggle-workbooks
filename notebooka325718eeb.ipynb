```python
import networkx as nx
G = nx.path_graph(1200)
pos = nx.spring_layout(G)
```

```python
import pandas as pd
df = pd.DataFrame().from_dict(data=pos, orient='index', columns=['x', 'y']).reset_index()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='x', y='y', hover_name='index')
```
