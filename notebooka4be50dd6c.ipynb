```python
import pandas as pd
train_df = pd.read_csv(filepath_or_buffer='/kaggle/input/customer-conversion-dataset-for-stuffmart-com/customer_conversion_traing_dataset .csv',
                      index_col=['LeadID'])
test_df = pd.read_csv(filepath_or_buffer='/kaggle/input/customer-conversion-dataset-for-stuffmart-com/customer_conversion_testing_dataset.csv',
                     index_col=['LeadID'])
df = pd.concat(objs=[train_df, test_df])
df.info()
```

```python
df.sample(n=10, random_state=2023)
```

```python
from plotly.express import histogram
for column in ['Age', 'TimeSpent (minutes)', 'PagesViewed', 'EmailSent', 'FormSubmissions', 'Downloads', 'CTR_ProductPage',
              'ResponseTime (hours)', 'FollowUpEmails', 'SocialMediaEngagement', ]:
    histogram(data_frame=df, x=column, color='Conversion (Target)').show()
```

```python
from plotly.express import scatter
scatter(data_frame=df.sample(n=10000), x='Age', y='TimeSpent (minutes)', color='Conversion (Target)')
```
