```python
import numpy as np
import pandas as pd

URBAN = '/kaggle/input/urban-population-analysis1950-2050/dataset.csv'

# we have an empty column and we have some N/A values; let's clean those up immediately 
df = pd.read_csv(filepath_or_buffer=URBAN).replace('NOT APPLICABLE', np.nan)
df = df.drop(columns=df.columns[4]).dropna()
# now all our numerical values can be floats
df[df.columns[3:]] = df[df.columns[3:]].astype(float)
df.head()
```

We really have four sets of data: individual countries or groups of countries, and value (in thousands) or percentage.

```python
df.columns
```

```python
from plotly.express import line
ys = ['Absolute value in thousands', 'Urban population as percentage of total population',]
for y in ys:
    line(data_frame=df[(df['Economy'] > 1000) | (df['Economy'] == 0)], x='Year', y=y, color='Economy Label', height=1200, log_y=(y==ys[0])).show()
```

```python
for y in ys:
    line(data_frame=df[(df['Economy'] > 0) | (df['Economy'] < 1000)], x='Year', y=y, color='Economy Label', height=1200, log_y=(y==ys[0])).show()
```
