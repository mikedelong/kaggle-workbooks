```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/apple-stock-price/AAPL.csv', parse_dates=['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

```python
from plotly.express import line
line(
    x='Date', y=['Close', 'Adj Close'], log_y=True,
    data_frame=df.set_index(keys=['Date']).resample('M').mean().reset_index(), 
    title='Weekly mean close/adjusted close AAPL price',
    )
```

We need to use a log plot here because the price has risen so much since 2003. 

```python
from plotly.express import scatter
scatter(data_frame=df, y='Volume', x='Adj Close', log_x=True, log_y=True, color='year', trendline='lowess')
```

Both price and volume are a function of time, but volume has first risen and then fallen as the price has risen.
