```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/sri-lankan-rupee-rhythms-13-year-forex-dance/exchange_rates_sl_dataset.csv', parse_dates=['Date'])
df.head()
```

```python
from plotly.express import line
buy_df = df[['Currency', 'Date', 'Buy Rate',]].pivot_table(columns=['Currency'], index=['Date']).reset_index().set_index(keys=['Date'])
buy_df.columns = buy_df.columns.get_level_values(1)
line(data_frame=buy_df, x=buy_df.index, y=df['Currency'].unique(), log_y=True)
```

```python
sell_df = df[['Currency', 'Date', 'Sell Rate',]].pivot_table(columns=['Currency'], index=['Date']).reset_index().set_index(keys=['Date'])
sell_df.columns = sell_df.columns.get_level_values(1)
line(data_frame=sell_df, x=sell_df.index, y=df['Currency'].unique(), log_y=True)
```
