```python
import pandas as pd
df = pd.read_csv('/kaggle/input/police-shootings/fatal-police-shootings-data.csv', index_col=['id'], parse_dates=['date'])
df.head()
```

```python
df.info()
```

We expect age to be a big driver so let's start looking at age histograms broken out by other factors.

```python
from plotly.express import histogram
for column in ['threat_type', 'flee_status', 'armed_with', 'state', 'gender', 'race', 'was_mental_illness_related', 'body_camera']:
    histogram(data_frame=df.sort_values(by=column, ascending=(column == 'state')), x='age', color=column, hover_name='name').show()
```
