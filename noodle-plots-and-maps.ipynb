```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/world-instant-noodles-consumption-2022/noodles.csv').drop(columns=['Country/Territory'])
df['per_million'] = 1e6 * df['2022'] / df['2022 Population']
df.head()
```

We should probably measure noodle consumption on a per capita basis since we have the population data.

```python
df.info()
```

```python
from plotly.express import choropleth
choropleth(data_frame=df, locations='CCA3', color='per_million')
```

```python
from plotly.express import bar
bar(data_frame=df.sort_values(ascending=False, by='per_million'), x='Country/Region', y='per_million', color='Rank',
   color_continuous_scale='Viridis_R')
```

This is probably the best way to view this data because it captures both the per-capita data and the gross consumption rank.

```python
from plotly.express import scatter
scatter(data_frame=df, x='2018', y='2022', hover_name='Country/Region', color='per_million', log_x=True, log_y=True,)
```

This just captures the change in population over the period of interest.
