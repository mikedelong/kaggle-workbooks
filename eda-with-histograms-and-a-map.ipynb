```python
import pandas as pd

NYC = '/kaggle/input/nyc-shooting-incidents-2006-2022/NYC_shooting_incidents.csv'

df = pd.read_csv(filepath_or_buffer=NYC, index_col=[0], parse_dates=['OCCUR_DATE'])

df.head()
```

Which borough is the most violent?

```python
from plotly import express

express.histogram(data_frame=df, x='BORO')
```

How are murder victims distributed according to their age?

```python
express.histogram(data_frame=df, x='VIC_AGE_GROUP')
```

We expect there to be more male than female victims, but how many more?

```python
express.histogram(data_frame=df, x='VIC_SEX')
```

We don't really have a victim racial expectation because we don't know the racial makeup of NYC, but let's look anyway.

```python
express.histogram(data_frame=df, x='VIC_RACE')
```

We expect to see a COVID effect on murders; do we see it and if so what does it look like?

```python
express.histogram(data_frame=df, x='YEAR')
```

We also expect to see within years some seasonality with murders peaking in the summer and declining in the winter primarily due to ambient temperature. NYC can get pretty cold in the winter.

```python
express.histogram(data_frame=df, x='MONTH')
```

Finally do murders have seasonality within an average day? Do we expect more murders in the daytime or at night?

```python
express.histogram(data_frame=df, x='HOUR')
```

Before we go let's make a map. We have lat/lon data so a map should be easy; how should we color our points? Let's pick age.

```python
import warnings
from plotly import express

warnings.filterwarnings(action='ignore', category=FutureWarning)
express.scatter_mapbox(mapbox_style='open-street-map', zoom=10, height=800, data_frame=df[df['YEAR'] > 2019].sample(frac=1, random_state=2024), lat='Latitude', lon='Longitude', color='VIC_AGE_GROUP', )
```
