```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/google-stock-price/GOOG.csv', parse_dates=['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'])
```

```python
line(data_frame=df.set_index(keys=['Date']).resample('M').mean().reset_index(), x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'],
    log_y=True)
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Adj Close', y='Volume', color='year', log_x=True, log_y=True, trendline='lowess')
```
