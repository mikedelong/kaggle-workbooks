```python
from arrow import now
from ctransformers import AutoModelForCausalLM

MAX_NEW_TOKENS = 2048
MODEL_PATH = 'TheBloke/Mistral-7B-Instruct-v0.1-GGUF'
# MODEL_FILE = 'mistral-7b-instruct-v0.1.Q4_K_M.gguf'
MODEL_FILE = 'mistral-7b-instruct-v0.1.Q5_K_S.gguf'
MODEL_TYPE = 'mistral'
TEMPERATURE = 0.7
THREADS = 4

setup_start = now()
model = AutoModelForCausalLM.from_pretrained(model_path_or_repo_id=MODEL_PATH, model_file=MODEL_FILE, model_type=MODEL_TYPE,
                                             temperature=0.7, gpu_layers=0, stream=False, threads=THREADS, reset=True,
                                             max_new_tokens=MAX_NEW_TOKENS, )
print('built model in {}'.format(now() - setup_start))
```

```python
model(prompt='Please classify these sentences: ["This restaurant is awesome.", "This restaurant is awful."] and give me a numerical value for how positive or negative each one is.')
```

```python
from arrow import now
time_start = now()
sentences = [
    'If I had a face like that I would wear a mask.',
    'Even animals have hopes and dreams.',
    'You are not going out in public dressed like that are you?',
]
for sentence in sentences:
    
    prompt = 'Please classify this sentence: [{}] as positive or negative and give me a numerical value of how positive or negative it is.'.format(sentence)
    print(model(prompt=prompt))
print('done in {}'.format(now() - time_start))
```
