```python
import pandas as pd
from chardet import detect    
filename='/kaggle/input/salaries-by-gender/week4_australian_salary.csv'
result = detect(open(file=filename, mode= 'rb').read())
df = pd.read_csv(filepath_or_buffer=filename, encoding=result['encoding'], index_col=[0])
df.head()
```

```python
df.info()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='average_taxable_income', color='gender', barmode='overlay')
```
