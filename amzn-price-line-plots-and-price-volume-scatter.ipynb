```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/amazon-com-inc-amzn/AMZN.csv', parse_dates=['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'], log_y=True)
```

AMZN has run up so much in 25 years we definitely need to look at the log of the price to see the dot com crash at all.

```python
from plotly.express import scatter
scatter(data_frame=df.set_index(keys=['Date']).resample('Q').mean().reset_index(), x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'],
        trendline='ols', trendline_scope='overall', log_y=False)
```

It is kind of surprising how much of AMZN's total gain can be explained with a linear model. The last five years do not look like linear growth at all.

```python
from plotly.express import scatter
scatter(data_frame=df, x='Adj Close', y='Volume', color='year', log_y=True, trendline='ols')
```

With the FAANG stocks generally we see this behavior, where when they were cheap on a per-share basis the volume was quiet high as a function of price, and as the price has increased volume has decreased. The obvious explanation would be that share price decreases volume, but another good explanation would be that volume is driven by the size of the float (the number of shares available for trading). We don't have float data in this dataset so we can't test that theory here.
