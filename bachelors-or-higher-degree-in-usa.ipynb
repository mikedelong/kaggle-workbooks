```python
import pandas as pd
df = pd.read_csv('../input/bachelors-or-higher-degree-data-of-usa/Bachelors or Higher Degree for USA States.csv').rename(columns={'Unnamed: 0': 'Year'})
df.rename(columns=lambda x: x.strip())
df['Year'] = pd.to_datetime(df['Year']).apply(lambda x: x.year)
df.head()
```

```python
from plotly.express import line 
figure = line(df, x='Year', y=df.columns[1:])
figure.show()
```
