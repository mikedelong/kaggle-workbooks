```python
import pandas as pd
filename = '../input/bachelors-or-higher-degree-data-of-usa/Bachelors or Higher Degree for USA States.csv'
df = pd.read_csv(filepath_or_buffer=filename).rename(columns={'Unnamed: 0': 'Year'})
df = df.rename(columns=lambda x: x.strip())
df['Year'] = pd.to_datetime(df['Year']).apply(lambda x: x.year)
df = df.set_index(keys=['Year'])
df.head()
```

```python
from pandas import DataFrame
from plotly.graph_objects import Figure 
from plotly.express import line 

def make_sorted_figure(arg_df: DataFrame) -> Figure:
    result_df = arg_df.reindex(arg_df.mean().sort_values(ascending=False,).index, axis=1).reset_index(drop=True)
    result = line(result_df, x=result_df.index, y=result_df.columns,)
    result.update_layout(xaxis_title='Year', yaxis_title='Rate')
    return result

make_sorted_figure(arg_df=df).show()
```

```python
relative_df = df.copy(deep=True)
for column in relative_df.columns:
    relative_df[column] = relative_df[column] / relative_df[column].values[0]
    
make_sorted_figure(arg_df=relative_df).show()
```
