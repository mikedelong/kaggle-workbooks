```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/bachelors-or-higher-degree-data-of-usa/Bachelors or Higher Degree for USA States.csv").rename(columns={'Unnamed: 0': 'Year'})
df.columns = [item.strip() for item in df.columns]
df['Year'] = df['Year'].apply(func=lambda x: int(str(x)[:4]))
df.head()
```

```python
%matplotlib inline
from matplotlib.pyplot import xticks
from seaborn import lineplot
lineplot(data=df.melt(id_vars='Year', value_vars=df.columns[1:], var_name='State'), x='Year', y='value', hue='State', legend=None, )
_ = xticks(rotation=90)
```
