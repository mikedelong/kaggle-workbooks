```python
import pandas as pd

df = pd.read_json(path_or_buf='/kaggle/input/hotel-rooms/deviation.json')
df.head()
```

```python
df.info()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='min', y='max', hover_name='name', log_x=True).show()
```

```python
from seaborn import heatmap
heatmap(data=df[df.columns[3:]].corr(numeric_only=True))
```

```python
from seaborn import pairplot
pairplot(data=df[df.columns[3:]], diag_kind='kde')
```
