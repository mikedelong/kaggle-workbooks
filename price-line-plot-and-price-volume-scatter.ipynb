```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/arm-stock-data-all-time/ARM.csv', parse_dates=['Date'])
df['month'] = df['Date'].dt.month
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'])
```

This curve is already pretty smooth, partly because we have a relatively short series.

```python
line(data_frame=df.set_index('Date').resample('W').mean().reset_index(), x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'])
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Volume', y='Adj Close', log_x=True, hover_name='Date', color='month', trendline='ols')
```

Prices and volume are very lightly correlated.
