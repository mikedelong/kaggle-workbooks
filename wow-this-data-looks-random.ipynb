```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/predict-lifespan-of-a-comet-goldfish/fish_data.csv', index_col=['id'])
df.head()
```

```python
from plotly.express import scatter
for color in ['habitat', 'ph_of_water', 'color', 'Gender', 'life_span']:
    scatter(data_frame=df, x='average_length(inches))', y='average_weight(inches))', color=color).show()
```

This data looks oddly random.

```python
from plotly.express import bar
for column in ['habitat', 	 	'color' ,	'Gender', 	]:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=2023, verbose=1)
df[['t0', 't1']] = tsne.fit_transform(X=df[['average_length(inches))', 	'average_weight(inches))', 	 	'ph_of_water' 	]])
scatter(data_frame=df, x='t0', y='t1', color='life_span')
```

The TSNE model clumps the cases together based on the numerical data, but the clumps don't seem to have similar life spans.
