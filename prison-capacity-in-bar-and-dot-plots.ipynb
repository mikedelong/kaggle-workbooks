```python
import pandas as pd
# drop columns that are always the same
df = pd.read_csv(filepath_or_buffer='/kaggle/input/prison-capacity-and-number-of-persons-held-in-eu/estat_crim_pris_cap_en.csv').drop(columns=['DATAFLOW', 'LAST UPDATE', 'freq', 'OBS_FLAG'])
df.head()
```

```python
df.nunique()
```

```python
from plotly.express import bar
for column in ['indic_cr', 'unit', 'geo', 'TIME_PERIOD']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```

```python
from plotly.express import scatter
for indicator in df['indic_cr'].unique():
    scatter(data_frame=df[df['indic_cr'] == indicator], x='geo', y='TIME_PERIOD', color='OBS_VALUE', facet_col='unit', facet_col_wrap=1, title=indicator).show()
```

We don't really have three numerical axes to use to plot our data; if we use the location and year we get this grid, and most of our capacities are relatively small.

```python
for indicator in df['indic_cr'].unique():
    scatter(data_frame=df[df['indic_cr'] == indicator], x='geo', color='TIME_PERIOD', y='OBS_VALUE', facet_col='unit', title=indicator, facet_col_wrap=1,
           log_y=True).show()
```

By using the time as the color we can see changes over time.
