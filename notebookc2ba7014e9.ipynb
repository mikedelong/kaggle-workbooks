```python
import pandas as pd
# category is always 'Notebooks' so we can drop it
df = pd.read_csv(filepath_or_buffer='/kaggle/input/laptops-and-notebooks-from-2020-2023/mindfactory_done.csv').drop(columns=['category'])
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
from plotly.express import scatter
scatter(data_frame=df.dropna(subset=['depth_mm']), x='width_mm', y='height_mm', hover_name='name', color='display_cm',
        size='depth_mm', opacity=0.2, log_y=True)
```

```python
scatter(data_frame=df.dropna(subset=['price_eur']), x='width_mm', y='height_mm', hover_name='name', color='weight_kg',
        size='price_eur', opacity=0.2, log_y=True)
```

We probably have some data quality issues with the weight data; I do no believe anyone is selling 900kg laptops; those things would weigh more than a motorcycle.

```python
from plotly.express import histogram
histogram(data_frame=df, x='weight_kg')
```

Let's remove our outliers.

```python
histogram(data_frame=df[df['weight_kg'] < 100], x='weight_kg')
```

```python
scatter(data_frame=df[df['weight_kg'] < 100].dropna(subset=['price_eur']), x='width_mm', y='height_mm', hover_name='name', color='weight_kg',
        size='price_eur', opacity=0.2, log_y=True)
```

```python
from plotly.express import bar
for column in ['storage_type', 'battery_life_h', 'battery_capacity_wh', 'psu_watts', 'audio_system', 'speakers_count',
               'has_touchscreen', 'keyboard_backlit', 'keyboard_numpad', 'has_webcam', 'has_bluetooth', 'bluetooth_version', 
               'wifi_standard', 'release_year']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```
