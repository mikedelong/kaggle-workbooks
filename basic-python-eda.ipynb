```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/ai-5000-tools-2023/all_ai_tool.csv')
df.head()
```

```python
df.nunique()
```

```python
df.info()
```

```python
from matplotlib.pyplot import subplots
_, ax = subplots(figsize=(14, 6), ncols=2)
for index, column in enumerate(['Free/Paid/Other', 'Major Category']):
    df[column].value_counts().nlargest(n=8).plot(ax=ax[index], kind='bar')
```

```python
from matplotlib.pyplot import axis
from matplotlib.pyplot import imshow
from matplotlib.pyplot import subplots
from wordcloud import WordCloud
from wordcloud import STOPWORDS
subplots(figsize=(14, 6))
axis('off')
imshow(
    X=WordCloud(stopwords=STOPWORDS | {'AI', 'tool', 'powered', 'data', 'create', 'platform', 'use'}).generate(text=' '.join(df['Description'].values.tolist())),
    cmap='viridis',
)
```
