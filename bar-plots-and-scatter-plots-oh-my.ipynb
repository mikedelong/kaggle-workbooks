```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/worlds-air-quality-and-water-pollution-dataset/cities_air_quality_water_pollution.18-10-2021 (1).csv')
df.columns = [item.replace('"', '').strip() for item in df.columns]
df['country'] = df['Country'].apply(lambda x: x.replace('"', '').strip())
df.head()
```

```python
df.info()
```

```python
from plotly.express import bar
bar(data_frame=df['country'].value_counts().to_frame().reset_index().sort_values(ascending=False, by='count').head(20), 
    x='country', y='count', log_y=True)
```

```python
from plotly.express import histogram
for column in df.columns[-3:-1]:
    histogram(data_frame=df, x=column, nbins=100).show()
```

Our two series of data are distributed rather differently; how are they correlated?

```python
from plotly.express import imshow
imshow(img=df[['WaterPollution', 'AirQuality']].corr())
```

They are negatively correlated; substantially so.

```python
from plotly.express import scatter
scatter(data_frame=df[df['country'].isin(df['country'].value_counts()[:10].index.tolist())], x='WaterPollution', y='AirQuality', height=600,
        color='country', title='Data for top 10 countries by datapoint count', hover_name='City')
```

This data looks like it might be of variable quality.

```python
bar(data_frame=df[['country', 'AirQuality']].groupby(by='country').mean().reset_index().sort_values(ascending=False, by='AirQuality').head(n=20), x='country', y='AirQuality', title='Top 20 countries by mean reported air quality')
```

```python
bar(data_frame=df[['country', 'WaterPollution']].groupby(by='country').mean().reset_index().sort_values(ascending=False, by='WaterPollution').head(n=20), x='country', y='WaterPollution', title='Top 20 countries by mean reported water pollution')
```
