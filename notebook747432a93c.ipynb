```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/diabetes/diabetes.csv', index_col=['id'])
df['diabetes'] = df['stab.glu'] > 125
df.head()
```

```python
from plotly.express import bar
for column in ['location', 'gender', 'frame', 'diabetes']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```

```python
from plotly.express import histogram
for column in ['chol', 'stab.glu', 'hdl', 'ratio', 'glyhb', 'age',  'height', 'weight', 'bp.1s', 'bp.1d', 'bp.2s', 'bp.2d', 'waist',
               'hip', 'time.ppn']:
    histogram(data_frame=df, x=column, color='diabetes').show()
```

Diabetes is our target variable, so it makes sense to color the histograms based on its value. It is defined in terms of stab.glu so its correlation with the other variables is what is interesting.

```python
df.info()
```

```python
from plotly.express import scatter
scatter(data_frame=df[df['chol'].notnull()], x='height', y='weight', color='gender', size='chol', symbol='frame')
```

```python
scatter(data_frame=df[df['chol'].notnull()], x='height', y='weight', color='diabetes', size='chol', symbol='frame')
```

```python
scatter(data_frame=df[df['chol'].notnull()], x='waist', y='hip', color='gender', size='chol', symbol='frame', opacity=0.5)
```

There's more dimorphism here than I would have expected. 

```python
scatter(data_frame=df[df['chol'].notnull()], x='waist', y='hip', color='diabetes', size='chol', symbol='frame', opacity=0.5)
```

```python
scatter(data_frame=df[df['weight'].notnull()], x='age', y='glyhb', color='diabetes', size='weight', symbol='gender', opacity=0.5)
```

```python
from plotly.express import imshow
imshow(img=df.corr(numeric_only=True))
```

```python
imshow(img=df[['chol', 'stab.glu', 'hdl', 'ratio', 'glyhb', 'age',  'height', 'weight', 'bp.1s', 'bp.1d', 'bp.2s', 'bp.2d', 'waist',
               'hip', 'time.ppn', 'diabetes']].corr()['diabetes'][:-1].to_frame().T)
```

```python
scatter(data_frame=df[['chol', 'stab.glu', 'hdl', 'ratio', 'glyhb', 'age',  'height', 'weight', 'bp.1s', 'bp.1d', 'bp.2s', 'bp.2d', 'waist',
               'hip', 'time.ppn', 'diabetes']].corr()['diabetes'][:-1].to_frame().reset_index().reset_index(), x='level_0', y='diabetes',
         hover_name='index', color='diabetes')
```

This is probably the nut graf; most of our variables that are not directly related to blood sugar are not highly correlated with the target variable.
