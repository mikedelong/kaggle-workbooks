```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/hollywood-age-gap/movies.csv', parse_dates=['Actor 1 Birthdate', 'Actor 2 Birthdate'])
df['gender'] = df['Actor 1 Gender'] + '/' + df['Actor 2 Gender']
df.head()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='Release Year')
```

Our dataset is dominated by recently-released movies.

```python
from plotly.express import pie
pie(data_frame=df, names='gender', color='gender')
```

Our dataset is dominated by one kind of couple.

```python
from plotly.express import bar
bar(data_frame=df['Director'].value_counts().nlargest(n=25).to_frame(),title='Top 25 directors',)
```

```python
from plotly.express import scatter
for color in ['gender', 'Release Year', 'Age Difference']:
    scatter(data_frame=df, x='Actor 1 Age', y='Actor 2 Age', color=color, hover_name='Movie Name', trendline='ols', trendline_scope='overall').show()
```
