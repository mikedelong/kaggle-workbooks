```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/starbucks-reviews-dataset/reviews_data.csv')
# let's consolidate different capitalizations of names
df['Name'] = df['name'].str.capitalize()
# let's try to get the state out of the location
to_replace = {'MINNESOTA': 'MN', 'ALBERTA': 'AL', 'MICHIGAN': 'MI', 'CALIFORNIA': 'CA', 'WI.': 'WI',
             'MAINE': 'ME'}
df['state'] = df['location'].apply(func=lambda x: x.upper().split()[-1]).replace(to_replace=to_replace)
df.head()
```

```python
df.info()
```

```python
from plotly.express import bar
bar(data_frame=df['Name'].value_counts().nlargest(n=20).to_frame())
```

```python
from plotly.express import treemap
treemap(data_frame=df['state'].value_counts().to_frame().reset_index(), path=['state'], values='count')
```

```python
treemap(data_frame=df['Name'].value_counts().to_frame().reset_index(), path=['Name'], values='count')
```

```python
bar(data_frame=df.sort_values(by='Rating'), x='state', color='Rating')
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='Rating')
```
