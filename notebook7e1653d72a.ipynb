```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/starbucks-reviews-dataset/reviews_data.csv')
# let's consolidate different capitalizations of names
df['Name'] = df['name'].str.capitalize()
# let's try to get the state out of the location
to_replace = {'MINNESOTA': 'MN', 'ALBERTA': 'AL', 'MICHIGAN': 'MI', 'CALIFORNIA': 'CA', 'WI.': 'WI',
             'MAINE': 'ME'}
df['state'] = df['location'].apply(func=lambda x: x.upper().split()[-1]).replace(to_replace=to_replace)
df.head()
```

```python
df.info()
```

```python
from plotly.express import bar
bar(data_frame=df['Name'].value_counts().nlargest(n=20).to_frame())
```

```python
from plotly.express import treemap
treemap(data_frame=df['state'].value_counts().to_frame().reset_index(), path=['state'], values='count')
```

```python
treemap(data_frame=df['Name'].value_counts().to_frame().reset_index(), path=['Name'], values='count')
```

```python
bar(data_frame=df.sort_values(by='Rating'), x='state', color='Rating')
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='Rating')
```

This dataset is a lot more negative than I would have guessed.

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from matplotlib.pyplot import axis
from matplotlib.pyplot import imshow
from matplotlib.pyplot import subplots
subplots(figsize=(12, 12))
axis('off')
imshow(WordCloud(random_state=2023,).generate(' '.join(df['Review'].values.tolist())))
```

Let's split the reviews into positive (rating > 3) and negative (rating < 3) and look at the word clouds.

```python
subplots(figsize=(12, 12))
axis('off')
imshow(WordCloud(random_state=2023).generate(' '.join(df[df['Rating'] > 3]['Review'].values.tolist())))
```

```python
subplots(figsize=(12, 12))
axis('off')
imshow(WordCloud(random_state=2023).generate(' '.join(df[df['Rating'] < 3]['Review'].values.tolist())))
```

Let's start removing high-frequency tokens to see if we can get the high and low rating reviews to distinguish themselves.

```python
stopwords = {
    'Starbucks', 'coffee', 'drink', 'store', 'order', 'customer', 'service', 
}
print(stopwords)
```

```python
subplots(figsize=(12, 12))
axis('off')
imshow(WordCloud(random_state=2023, stopwords=STOPWORDS | stopwords).generate(' '.join(df[df['Rating'] > 0]['Review'].values.tolist())))
```

```python
from matplotlib.pyplot import figure
fig = figure(figsize=(12, 12))
for i in range(2):
    ax = fig.add_subplot(2, 1, i+1)
    if i == 0:
        wordcloud = WordCloud(random_state=2023, stopwords=STOPWORDS | stopwords).generate(' '.join(df[df['Rating'] < 3]['Review'].values.tolist()))
    else:
        wordcloud = WordCloud(random_state=2023, stopwords=STOPWORDS | stopwords).generate(' '.join(df[df['Rating'] > 3]['Review'].values.tolist()))
    ax.imshow(wordcloud)
    ax.axis('off')
```

It doesn't take much for the low and high rated reviews to distinguish themselves.
