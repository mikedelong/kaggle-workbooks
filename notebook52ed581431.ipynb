```python
filename = '/kaggle/input/viral-shorts-youtubes-most-viewed/most_viewed_youtube_shorts.csv'
```

```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer=filename, parse_dates=['published_at'])
df['seconds'] = df['duration'].apply(func=lambda x: pd.Timedelta(x).total_seconds())
df.head()
```

```python
df.shape
```

```python
df.nunique()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='published_at')
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='published_at', y='view_count', hover_name='title', log_y=True)
```

Does this match what we would expect? We would expect that if we measure virality in total views that views would grow over a period of time and then stabilize. So we would expect most of the viral videos would be older and the new ones would still be accumulating views. The data above still looks odd, but it is kind of understandable.
