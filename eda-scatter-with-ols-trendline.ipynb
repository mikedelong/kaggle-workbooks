```python
import pandas as pd

ICECREAM = '/kaggle/input/temperature-and-ice-cream-sales/Ice Cream Sales - temperatures.csv'
df = pd.read_csv(filepath_or_buffer=ICECREAM)
df.head()
```

We just have two columns so let's plot our data and see what they look like.

```python
from plotly import express
express.scatter(data_frame=df, x='Temperature', y='Ice Cream Profits', trendline='ols')
```

Our data is essentially linear: the R2 for our OLS trendline is 0.98. Obviously without other data we don't have much else to say about this data.
