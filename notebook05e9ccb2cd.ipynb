```python
import pandas as pd
from glob import glob
file_glob = glob('/kaggle/input/alberta-wildlife-watch-animal-carcass-records/open-data_aww-*.csv')
df = pd.concat(objs=[pd.read_csv(filepath_or_buffer=input_file) for input_file in file_glob])
df.columns = [column.strip() for column in df.columns]
df.info()
```

```python
df.head()
```

```python
from matplotlib.pyplot import subplots
_, ax = subplots(figsize=(14, 6), ncols=2)
for index in range(2):
    df['Species'].value_counts().plot(ax=ax[index], kind='bar', log=(index == 1))
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Latitude', y='Longitude', hover_name='Species', color='Year Observed').show()
```
