```python
import pandas as pd

DATA = '/kaggle/input/turbo-az-car-sales-dataset-in-azerbaijan/Turbo.az_car_data.xlsx'
sheet_name = 'Sheet1'

df = pd.read_excel(io=DATA, sheet_name=sheet_name)
df.head()
```

We have a naive expectation as to how our numerical values correlate, so let's correlate the numerical data.

```python
from plotly import express

express.imshow(img=df.corr(numeric_only=True))
```

Here's what we see:
* Newer cars are generally more expensive than older cars, because for the most part cars depreciate, so year and price are positively correlated
* Older cars have more miles on them, so year and kilometers are negatively correlated
* Cars mostly depreciate, so mileage (in km) and price are negatively correlated

This is probably the nut graf, and everything below is detail and commentary.

```python
import warnings
from plotly import express

warnings.filterwarnings(action='ignore', category=FutureWarning)
express.scatter(data_frame=df, x='year', y='price_usd', color='age_category', log_y=True, height=800, hover_name='name')
```

We see some older cars with higher prices, but they are anomalies, outliers, and luxury/rare cars. The broad trend is for cars to depreciate as they get older.

```python
express.scatter(data_frame=df, x='engine_power (cm3)', y='km', color='price_usd', height=800)
```

Engine power and mileage are uncorrelated, and to a degree engine power in cubic centimeters is a discrete variable.
