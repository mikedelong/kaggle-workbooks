```python
import pandas as pd

UNICORNS = '/kaggle/input/all-the-unicorns-in-the-world/List of Unicorns in the World.csv'

df = pd.read_csv(filepath_or_buffer=UNICORNS, index_col=[0], parse_dates=['Date Joined'])
df['valuation'] = df['Valuation ($B)'].apply(func=lambda x: float(x.replace('$', '')))
continents_df = pd.read_csv(filepath_or_buffer='/kaggle/input/country-mapping-iso-continent-region/continents2.csv', usecols=['name', 'alpha-3'])
df.head()
```

```python
import warnings
from plotly import express

warnings.filterwarnings(action='ignore', category=FutureWarning)
express.strip(data_frame=df, x='Date Joined', y='valuation', color='Country', log_y=True, height=1200, hover_name='Company')
```

```python
express.choropleth(data_frame=df.merge(right=continents_df, left_on='Country', right_on='name')[['alpha-3', 'valuation']].groupby(by='alpha-3').sum().reset_index(), locations='alpha-3', color='valuation', title='Total unicorn valuation ($Bn)').show()
express.choropleth(data_frame=df.merge(right=continents_df, left_on='Country', right_on='name')[['alpha-3', ]].groupby(by='alpha-3').size().reset_index(), locations='alpha-3', color=0, title='Unicorn count').show()
express.choropleth(data_frame=df.merge(right=continents_df, left_on='Country', right_on='name')[['alpha-3', 'valuation']].groupby(by='alpha-3').mean().reset_index(), locations='alpha-3', color='valuation', title='Mean unicorn valuation ($Bn)').show()
```
