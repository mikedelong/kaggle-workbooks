Let's load up our data and do just a little feature engineering. We know from the data card that the label class as 1 for male and 0 for female, so let's add a boolean column called Male.

```python
import pandas as pd

DATA = '/kaggle/input/vocal-gender-features/vocal_gender_features_new.csv'
df = pd.read_csv(filepath_or_buffer=DATA)
df['Male'] = df['label'] == 1
df.head()
```

```python
COLUMNS = ['mean_spectral_centroid', 'std_spectral_centroid',
       'mean_spectral_bandwidth', 'std_spectral_bandwidth',
       'mean_spectral_contrast', 'mean_spectral_flatness',
       'mean_spectral_rolloff', 'zero_crossing_rate', 'rms_energy',
       'mean_pitch', 'min_pitch', 'max_pitch', 'std_pitch', 'spectral_skew',
       'spectral_kurtosis', 'energy_entropy', 'log_energy', 'mfcc_1_mean',
       'mfcc_1_std', 'mfcc_2_mean', 'mfcc_2_std', 'mfcc_3_mean', 'mfcc_3_std',
       'mfcc_4_mean', 'mfcc_4_std', 'mfcc_5_mean', 'mfcc_5_std', 'mfcc_6_mean',
       'mfcc_6_std', 'mfcc_7_mean', 'mfcc_7_std', 'mfcc_8_mean', 'mfcc_8_std',
       'mfcc_9_mean', 'mfcc_9_std', 'mfcc_10_mean', 'mfcc_10_std',
       'mfcc_11_mean', 'mfcc_11_std', 'mfcc_12_mean', 'mfcc_12_std',
       'mfcc_13_mean', 'mfcc_13_std',]
RANDOM_STATE = 2025
TARGET = 'Male'
```

First let's use dimensionality reduction and make a big scatter plot of all of our data, colored by the target variable.

```python
from sklearn.manifold import TSNE


tsne = TSNE(random_state=RANDOM_STATE,)
tsne_df = pd.DataFrame(data=tsne.fit_transform(X=df[COLUMNS]), columns=['tx', 'ty'])
tsne_df['Male'] = df['Male'].tolist()
```

```python
from plotly import express

express.scatter(data_frame=tsne_df, x='tx', y='ty', color='Male')
```

What do we see? We see a lot of local clustering, but also a fair amount of mixing. We should probably have moderate expectations regarding how accurate a simple model can be.
