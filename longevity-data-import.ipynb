```python
import pandas as pd
df = pd.read_csv('/kaggle/input/life-longevity-factors/LiveLongerData.csv', index_col=['ID'])
df['sexes affected'] = df['sexes affected'].replace({'male': 'Male'})
df.head()
```

```python
from plotly.express import scatter

scatter(data_frame=df.reset_index(), x='ID', y='Years gained / lost', hover_name='Factor', color='sexes affected',
       size='strength of science as a number').show()
```
