```python
import pandas as pd
with pd.ExcelFile(path_or_buffer='/kaggle/input/behavioral-classes-brazilian-adolescents/Behavioral classes dataset of Brazilian adolescents/Dataset.xlsx') as workbook:
    df = pd.read_excel(io=workbook, sheet_name='Data',
                        index_col=[0]
                      )
df['sex'] = df['Sex'].map(arg={1: 'Female', 2: 'Male'}) # just guessing here
df['sleep_duration'] = df['SD'].apply(func=lambda x: x.hour + x.minute / 60)
df.head()
```

```python
df.info()
```

```python
from plotly.express import imshow
imshow(img=df.corr(numeric_only=True))
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Steps/day', y='BMI', color='sex', trendline='ols', )
```
