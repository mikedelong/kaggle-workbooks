```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/antidoping-company-usadatest/Usadatest2023.csv')
counts = df[['Sport', 'Sample Count']].groupby(by='Sport').max().to_dict()
df['Sport max Sample Count'] = df['Sport'].apply(func=lambda x: counts['Sample Count'][x])
df.head()
```

```python
df.nunique()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='Sample Count', log_y=True)
```

```python
df['Sample Count'].mean(), df['Sample Count'].median()
```

An obvious research question here would be to understand why if the mean/median are so close to zero why the upper quantile has athletes with so many samples.

```python
from plotly.express import scatter
scatter(data_frame=df.sort_values(by='Sport max Sample Count'), x='Sample Count', y='Sport', color='Sample Count', hover_name='Athlete Name', height=1500)
```
