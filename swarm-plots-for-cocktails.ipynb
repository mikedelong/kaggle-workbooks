```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/cocktailcrazedataset/data (1).csv', index_col=['idDrink']).drop(columns=['Unnamed: 0'])
df.head()
```

```python
from plotly.express import bar
for column in ['strCategory', 'strAlcoholic', 'strGlass', 'initial']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```

```python
from plotly.express import strip
strip(data_frame=df.sort_values(by='strGlass'), x='initial', y='strGlass', color='strCategory', height=800, hover_name='strDrink', 
      hover_data=['strAlcoholic'])
```

We have some duplicates here so the swarm/strip plot does a reasonable job.
