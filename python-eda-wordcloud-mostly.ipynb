```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/hospital-medical-facility/output.csv', index_col=[0])
df['District'] = df['District'].apply(func=lambda x: ' '.join([token.capitalize() for token in x.split()]))
df.head()
```

```python
df.nunique()
```

```python
from plotly.express import bar
for column in ['District', 'Clusters in HMIS application']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```

```python
from matplotlib.pyplot import subplots
from matplotlib.pyplot import axis
from matplotlib.pyplot import imshow
from wordcloud import WordCloud
subplots(figsize=(12, 12))
imshow(X=WordCloud(random_state=2023, height=1200, width=1200 ).generate(text=' '.join(df['Name'].values), ))
axis('off')
```

```python
subplots(figsize=(12, 12))
imshow(X=WordCloud(random_state=2023, height=1200, width=1200 ).generate(text=' '.join(df['Address of Hospital'].dropna().values), ))
axis('off')
```
