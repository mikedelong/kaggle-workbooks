```python
import pandas as pd
from glob import glob

df = pd.concat([pd.read_csv(filepath_or_buffer=filename) for filename in glob(pathname='/kaggle/input/drc-news-corpus/*.csv')]).drop_duplicates(ignore_index=True).drop(columns=['link'])
df.head()
```

```python
df.info()
```

```python
from plotly.express import bar
bar(data_frame=df[['categories', 'source']].groupby(by=['categories', 'source']).size().reset_index(), x='source', y=0, color='categories').show()
bar(data_frame=df[['categories', 'source']].groupby(by=['categories', 'source']).size().reset_index(), color='source', y=0, x='categories').show()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='timestamp', log_y=True)
```
