```python
import pandas as pd

DATA = '/kaggle/input/bank-employee-dataset/mydata.csv'

df = pd.read_csv(filepath_or_buffer=DATA)

df.head()
```

```python
df.info()
```

All of our data is numeric, so we can use dimension reduction to visualize using the entire set.

```python
import arrow
from plotly import express
from umap import UMAP

COLUMNS = [item for item in df.columns if item != 'deposit_yes']
TARGET = 'deposit_yes'

time_start = arrow.now()
reducer = UMAP(n_components=2, random_state=2024, transform_seed=2024, verbose=True, n_jobs=1, n_epochs=1000)
umap_df = pd.DataFrame(data=reducer.fit_transform(X=df[COLUMNS]), columns=['x', 'y'])
umap_df[TARGET] = df.dropna()[TARGET].tolist()
express.scatter(data_frame=umap_df, x='x', y='y', color=TARGET).show()
print('UMAP done in {}'.format(arrow.now() - time_start))
```
