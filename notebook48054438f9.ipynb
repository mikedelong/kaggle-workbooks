```python
import pandas as pd
dataset_df = pd.read_csv(filepath_or_buffer='/kaggle/input/international-iq-by-countries/International_IQ.csv', index_col=['Ranking'])
dataset_df['Country'] = dataset_df['Country'].replace(to_replace={'Bolivia (Plurinational State of)': 'Bolivia',
                                                                  'Bosnia and Herzegovina': 'Bosnia And Herzegovina',
                                                                  'Congo (Democratic Republic of the)': 'Congo (Democratic Republic Of The)',
                                                                  "Cote d'Ivoire": "CÃ´te D'Ivoire",'Czechia' : 'Czech Republic',
                                                                  'Iran (Islamic Republic of)' : 'Iran',
                                                                  "Lao People's Democratic Republic" : 'Laos','Moldova (Republic of)': 'Moldova',
                                                                  'North Macedonia':  'Macedonia', 'Russian Federation' : 'Russia', 
                                                                  'Venezuela (Bolivarian Republic of)' : 'Venezuela','Viet Nam': 'Vietnam',})
df = dataset_df.merge(right=pd.read_csv(filepath_or_buffer='/kaggle/input/country-mapping-iso-continent-region/continents2.csv', 
                                        usecols=['name', 'alpha-3', 'region']).rename(columns={'name': 'Country'}), on='Country')
df.head()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Participants', y='Average IQ', hover_name='Country', log_x=True, color='region')
```

We need to use a log plot because there's not much variation in the y direction and a lot in the x direction.

```python
from plotly.express import histogram
histogram(data_frame=df, x='Participants', color='region')
```

It might be reasonable to expect that the (many) samples in the big stack there at the left are low-resolution.

```python
histogram(data_frame=df, x='Average IQ', color='region')
```

If we look at the IQ data using a histogram the region/IQ clustering doesn't look so pronounced as it did in the scaterplot, but neither does it look random.

```python
from plotly.express import choropleth
choropleth(data_frame=df, locations='alpha-3', color='Average IQ')
```
