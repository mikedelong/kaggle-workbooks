```python
import pandas as pd 
df = pd.read_csv('/kaggle/input/most-popular-applications-from-google-play-store/MOST_POPULAR_APPS_FROM_GOOGLE_PLAY.CSV')
df.sample(10)
```

```python
# almost everthing is com, and the non-com packages are mostly ones and twos
# the package name top level has a long thin tail; we can roll it up into com and other
df['top_level'] = df['package_name'].apply(lambda x: str(x).split('.')[0])
df['com_or_not'] = df['top_level'].apply(lambda x: x if x == 'com' else 'not')
df['com_or_not'].value_counts().plot(kind='barh')
```
