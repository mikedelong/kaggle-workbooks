```python
import pandas as pd
filename = '/kaggle/input/earthquakes-2023-global/earthquakes_2023_global.csv'
df = pd.read_csv(filepath_or_buffer=filename, parse_dates=['time', 'updated'], )
df.head()
```

```python
df.nunique().to_frame().T
```

```python
from plotly.express import histogram
from plotly import io

io.renderers.default = 'iframe'
histogram(data_frame=df, x='magType').show()
```
