```python
with open(file='/kaggle/input/trump-classified-documents-indictment/US-v-Trump-Nauta-De-Oliveira-23-80101.txt', mode='r') as input_fp:
    text = input_fp.readlines()
```

```python
len(text)
```

```python
from matplotlib.pyplot import axis
from matplotlib.pyplot import imshow
from matplotlib.pyplot import show
from wordcloud import WordCloud
imshow(WordCloud().generate(' '.join(text)), interpolation='bilinear')
axis('off')
show()
```

```python
imshow(WordCloud().generate(' '.join(text).lower()), interpolation='bilinear')
axis('off')
show()
```
