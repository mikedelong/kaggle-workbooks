```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/years-of-schooling-worldwide/schooling.csv')
df['mean'] = df[df.columns[7:38]].mean(axis=1)
df['max'] = df[df.columns[7:38]].max(axis=1)
df['stdev'] = df[df.columns[7:38]].std(axis=1)
df.head()
```

```python
from plotly.express import choropleth
for color in ['Continent', 'Hemisphere', 'Human Development Groups', 'UNDP Developing Regions', 'HDI Rank (2021)',
             'max', 'mean', 'stdev']:
    choropleth(data_frame=df, locations='ISO3', color=color).show()
```

```python
from plotly.express import scatter
for color in ['Continent', 'Hemisphere', 'Human Development Groups', 'UNDP Developing Regions', 'HDI Rank (2021)', 'max',  ]:
    scatter(data_frame=df, x='mean', y='stdev', color=color).show()
```
