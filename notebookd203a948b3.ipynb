```python
import pandas as pd

DATA = '/kaggle/input/hemoglobin-data/Hemoglobin Data.csv'

df = pd.read_csv(filepath_or_buffer=DATA, index_col=['image_name'])
df.head()
```

We don't have a lot of data; let's use UMAP to do some dimension reduction and make a plot.

```python
import arrow
from plotly import express
from umap import UMAP

COLUMNS = ['mid_b_mean', 'mid_g_mean', 'mid_r_mean', 'mid_b_median', 'mid_g_median', 'mid_r_median', 'mid_l_mean', 'mid_a_mean', 'mid_br_mean', 'mid_l_median', 'mid_a_median', 'mid_br_median', 'mid_h_mean', 'mid_s_mean', 'mid_v_mean', 'mid_h_median', 'mid_s_median', 'mid_v_median']
TARGET = 'hbvalues'
time_start = arrow.now()
reducer = UMAP(n_components=2, random_state=2024, transform_seed=2024, verbose=True, n_jobs=1, n_epochs=1000)
umap_df = pd.DataFrame(data=reducer.fit_transform(X=df[COLUMNS]), columns=['x', 'y'])
umap_df[TARGET] = df[TARGET].tolist()
express.scatter(data_frame=umap_df, x='x', y='y', color=TARGET, ).show()
print('UMAP done in {}'.format(arrow.now() - time_start))
```
