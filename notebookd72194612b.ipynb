```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/us-automatic-traffic-recorder-stations-data/Automatic_Traffic_Recorder_ATR_Stations.csv',
                index_col=[0]).drop(columns=['VERSION', 'X', 'Y'])
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
from plotly.express import histogram
for column in ['NHS', 'CONFIDENCE', 'YEAR_GEO', 'FCLASS', ]:
    histogram(data_frame=df, x=column).show()
```

```python
histogram(data_frame=df.sort_values(by='STPOSTAL'), x='STPOSTAL')
```

```python
for column in ['YEAR_GEO', ]:
    histogram(data_frame=df[df[column] > 0], x=column).show()
```

```python
from plotly.express import scatter_geo
scatter_geo(data_frame=df[df['YEAR_GEO'] > 0], locationmode='USA-states', projection='albers usa',
            lat='LATITUDE', lon='LONGITUDE', color='YEAR_GEO', hover_name='LOCATION')
```
