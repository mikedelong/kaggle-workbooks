```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
filename = '../input/dewey-decimal/DDS - Sheet1.csv'
df = pd.read_csv(filepath_or_buffer=filename, names=['class'], dtype={0: str})
df = df.sort_values(by='class').reset_index(drop=True)
# the one-character and two-charactter forms are aggregates so let's get them
df['one-character'] = df['class'].astype(str).str[:1]
df['two-character'] = df['class'].astype(str).str[:2]
df.head()
```

```python
%matplotlib inline
df['one-character'].hist(grid=False,)
```

```python
# Let's look at the two-character bins
df['two-character'].hist(bins=df['two-character'].astype(int).max() - df['two-character'].astype(int).min() + 1, grid=False,)
```

```python
# let's look at the class analog first even though it is ugly
df['class'].hist(bins=df['class'].astype(int).max() - df['class'].astype(int).min() + 1, grid=False, xrot=90, )
```

```python
# this probably looks better as a scatter plot
counts = df['class'].astype(int).value_counts()
pd.DataFrame(data={'class': counts.index, 'count': counts.values}).plot.scatter(x='class', y='count',)
```
