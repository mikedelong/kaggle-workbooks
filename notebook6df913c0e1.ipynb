```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
filename = '../input/dewey-decimal/DDS - Sheet1.csv'
df = pd.read_csv(filepath_or_buffer=filename, names=['class'], dtype={0: str})
df['one-character'] = df['class'].astype(str).str[:1]
df['two-character'] = df['class'].astype(str).str[:2]
df.head()
```

```python
%matplotlib inline
df['one-character'].sort_values().hist()
```

```python
bin_count = df['two-character'].astype(int).max() - df['two-character'].astype(int).min() + 1
df['two-character'].sort_values().hist(bins=bin_count)
```

```python
class_count = df['class'].astype(int).max() - df['class'].astype(int).min() + 1
df['class'].sort_values().hist(bins=class_count, grid=False, xrot=90, )
```

```python
counts = df['class'].astype(int).value_counts()
pd.DataFrame(data={'class': counts.index, 'count': counts.values}).plot.scatter(x='class', y='count',)
```
