```python
import pandas as pd
dataset_df = pd.read_csv(filepath_or_buffer='/kaggle/input/cats-vs-dogs/cats_vs_dogs.csv', index_col=['Unnamed: 0'])
df = dataset_df.merge(right=pd.read_csv(filepath_or_buffer='/kaggle/input/state-abbreviations/state_abbrev.csv', usecols=['State', 'Abbreviation']).rename(columns={'State': 'state'}), on='state', how='inner')
df.head()
```

```python
df.info()
```

```python
from plotly.express import bar
bar(data_frame=df.sort_values(ascending=False, by='n_households'), x='state', y='n_households',)
```

This looks a lot like a population map.

```python
from plotly.express import choropleth
choropleth(data_frame=df, locations='Abbreviation', scope='usa', color='percent_pet_households', locationmode='USA-states', title='Percentage of households with pets')
```

Weird we don't have data for Alaska or Hawaii. Maybe this is data from 1958 or from a possible future where Alaska and Hawaii are not states any more.

```python
for case in [('avg_cats_per_household', 'Cats per household', 'Reds'), ('avg_dogs_per_household', 'Dogs per household', 'Blues'),]:
    choropleth(data_frame=df, locations='Abbreviation', scope='usa', color=case[0], locationmode='USA-states', title=case[1],
               color_continuous_scale=case[2]).show()
```

Cats are girls and dogs are boys; I don't make the rules.

```python
for case in [('percent_cat_owners', 'Percent cat owners', 'Reds'), ('percent_dog_owners', 'Percent dog owners', 'Blues'),]:
    choropleth(data_frame=df, locations='Abbreviation', scope='usa', color=case[0], locationmode='USA-states', title=case[1],
               color_continuous_scale=case[2]).show()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='percent_cat_owners', y='percent_dog_owners', color='percent_pet_households', size='percent_pet_households',
       hover_name='state', trendline='ols')
```

This chart does a better job of expressing what an outlier DC is than the maps, because DC doesn't appear on the maps above at all. But that trend line probably captures the wrong thing; what we really want is a line that in a sense separates cat states from dog states, and our OLS trendline doesn't really do that, because it really captures a kind of central tendency.
