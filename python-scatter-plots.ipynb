```python
import pandas as pd
from numpy import log10
df = pd.read_csv(filepath_or_buffer='/kaggle/input/all-book-subject-1965-2022/All Book Subject Overview.csv')
for column in ['books', 'authors']:
    df['log10_' + column] = log10(df[column])
df.head()
```

```python
df.shape
```

```python
df[df['min_publication_date'] > 1500]['min_publication_date'].plot.hist(bins=200, log=True)
```

```python
from seaborn import pairplot
pairplot(data=df[['min_publication_date', 'max_publication_date', 'avg_publication_date']], diag_kind='kde')
```

```python
from seaborn import scatterplot
scatterplot(data=df.sample(100001), x='log10_books', y='log10_authors', )
```
