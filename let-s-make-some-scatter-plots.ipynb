```python
import pandas as pd

MODELS = '/kaggle/input/notable-ai-models-2024/notable_ai_models.csv'

df = pd.read_csv(filepath_or_buffer=MODELS, parse_dates=['Publication date'])
df.head()
```

How complete is our dataset? How much numeric data do we have that we can visualize?

```python
df.info()
```

Everybody talks about model size in terms of parameters, so let's graph those.

```python
from plotly import express

express.scatter(data_frame=df, x='Publication date', y='Parameters', hover_name='System', color='Domain', height=800, log_y=True)
```

We definitely see models getting larger over time, which is not surprising. Our dataset is kind of messy, and we don't have a good way to characterize multimodel models.

How about citation counts?

```python
express.scatter(data_frame=df, x='Publication date', y='Citations', hover_name='System', log_y=True)
```

Citation counts grow with time, but not all models accumulate citations over time.

```python
express.histogram(data_frame=df, x='Confidence')
```

We don't know a lot about these models, which is not surprising given that there are so many of them and they include so many details and probably trade secrets.
