```python
import pandas as pd
# drop columns with only one unique value
df = pd.read_csv(filepath_or_buffer='/kaggle/input/us-senate-election-returns-1976-2018/1976-2018-senate', 
                 sep='\t').drop(columns=['office', 'district', 'mode'])
df.head()
```

```python
from plotly.express import treemap
treemap(data_frame=df['party'].value_counts().to_frame().reset_index(), names='party', values='count', path=['party'],)
```

```python
from plotly.express import treemap
for year in df['year'].unique():
    treemap(data_frame=df[df['year'] == year]['party'].value_counts().to_frame().reset_index(), names='party', values='count', path=['party'],
           title='Party/candiate breakdown for {}'.format(year), height=400).show()
```

```python
from plotly.express import line
parties = ['republican', 'democrat', 'libertarian', 'independent', 'green', 'socialist workers']
line(data_frame=df[df['party'].isin(parties)], x='year', y='candidatevotes', color='state', height=800, log_y=True, facet_col='party').show()
```

```python
from plotly.express import bar
for column in ['stage', 'special', 'writein', 'unofficial', 'version']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count', height=400).show()
```
