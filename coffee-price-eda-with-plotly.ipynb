```python
import pandas as pd

COFFEE = '/kaggle/input/coffee-sales/index.csv'

df = pd.read_csv(filepath_or_buffer=COFFEE, parse_dates=['date', 'datetime'])
df.head()
```

```python
df.info()
```

Let's make a big scatter plot and see if anything emerges.

```python
import warnings
from plotly import express

warnings.filterwarnings(action='ignore', category=FutureWarning)
express.scatter(data_frame=df, x='datetime', y='money', color='coffee_name', symbol='cash_type')
```

From this it looks like prices went down in April and there appears to be a surcharge for cash, which is odd.

```python
express.bar(data_frame=df.drop(columns=['date', 'datetime', 'card']).groupby(by=['cash_type', 'coffee_name']).mean().reset_index(), 
            barmode='group', x='coffee_name', y='money', color='cash_type')
```

That is odd; maybe the cards are prepaid and we are giving a discount for using the prepaid card. Otherwise it makes more sense for there to be a surcharge for credit cards.
