```python
import arrow
import numpy as np
import pandas as pd

TEST = '/kaggle/input/digit-recognizer/test.csv'
TRAIN = '/kaggle/input/digit-recognizer/train.csv'

time_start = arrow.now()
test_df = pd.read_csv(filepath_or_buffer=TEST)
pixels = [column for column in test_df.columns if column.startswith('pixel')]
test_df['n0'] = test_df[pixels].apply(axis=1, func=lambda x: x.to_numpy(), )
test_df['n1'] = test_df['n0'].apply(func=lambda x: x.reshape(28, 28).transpose().reshape(-1))
test_df['value'] = test_df[['n0', 'n1']].apply(axis=1, func=np.concatenate)
train_df = pd.read_csv(filepath_or_buffer=TRAIN)
train_df['n0'] = train_df[pixels].apply(axis=1, func=lambda x: x.to_numpy(), )
train_df['n1'] = train_df['n0'].apply(func=lambda x: x.reshape(28, 28).transpose().reshape(-1))
train_df['value'] = train_df[['n0', 'n1']].apply(axis=1, func=np.concatenate)
print('{}: data load complete.'.format(arrow.now() - time_start))
```

```python
import arrow
from plotly import express
from umap import UMAP

TARGET = 'label'
time_start = arrow.now()
umap = UMAP(random_state=2024, verbose=True, n_jobs=1, low_memory=False, n_epochs=100)

plot_train_df = pd.DataFrame(columns=['x', 'y'], data=umap.fit_transform(X=train_df['value'].apply(pd.Series)))
plot_train_df[TARGET] = train_df[TARGET].tolist()
express.scatter(data_frame=plot_train_df, x='x', y='y', color=TARGET).show()
print('done with UMAP in {}'.format(arrow.now() - time_start))
```
