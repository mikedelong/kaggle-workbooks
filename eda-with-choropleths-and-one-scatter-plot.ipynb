```python
import pandas as pd

COUNTRIES = '/kaggle/input/best-country-to-live-in-2024/best-countries-to-live-in-2024.csv'
df = pd.read_csv(filepath_or_buffer=COUNTRIES)
df = df.merge(right=pd.read_csv(filepath_or_buffer='/kaggle/input/country-mapping-iso-continent-region/continents2.csv',
                               usecols=['name', 'alpha-3']),
             left_on='country', right_on='name').drop(columns=['name', 'share_borders'])
df.head()
```

```python
from plotly.express import choropleth
for color in [column for column in df.columns.tolist() if column not in {'country', 'alpha-3'}]:
    choropleth(data_frame=df, locations='alpha-3', color=color).show()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='population_growthRate', y='WorldHappiness2022', color='region', hover_name='country', log_x=True, log_y=True)
```

There isn't much to plot on a scatter plot that is interesting here, but plotting population growth against world happiness is kind of interesting.
