```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/worlds-100000-top-scientists/Top_scientists_2021.csv', index_col=['rank (ns)'])
df['ISO-3'] = df['cntry'].str.upper()
df.info()
```

```python
country_counts_df = df['ISO-3'].value_counts().to_frame().reset_index()
```

```python
df.head()
```

```python
from plotly.express import histogram
for x in ['c', 'h21']:
    histogram(data_frame=df, x=x).show()
```

We have a bunch of productivity metrics and they have different distributions.

```python
histogram(data_frame=df, x='self%').show()
```

Imagine being a top scientist and self-citing more than 90%

```python
df['self%'].nlargest(n=10)
```

```python
from plotly.express import imshow
imshow(img=df.corr(numeric_only=True), height=1000)
```

Not surprisingly our various productivity metrics are highly correlted, as seen by the hot-colored off-diagonal blocks above.

```python
from plotly.express import choropleth
choropleth(data_frame=country_counts_df, locations='ISO-3', color='count', scope='world', locationmode='ISO-3')
```

This map does not do a good job of representing the data; let's try a bar chart.

```python
from plotly.express import bar
bar(data_frame=country_counts_df.head(n=20), x='ISO-3', y='count')
```

```python
field_counts_df = df['sm-field'].value_counts().to_frame().reset_index()
bar(data_frame=field_counts_df, x='sm-field', y='count')
```
