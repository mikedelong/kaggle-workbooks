```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/worlds-100000-top-scientists/Top_scientists_2021.csv')
df['ISO-3'] = df['cntry'].str.upper()
df.info()
```

```python
country_counts_df = df.value_counts().to_frame().reset_index()
```

```python
from plotly.express import choropleth
choropleth(data_frame=country_counts_df, locations='ISO-3', color='count').show()
```
