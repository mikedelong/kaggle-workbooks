```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/material-footprint-per-capita-by-country/material_footprint.csv')
# if we don't have data in the most complete year we may as well drop the country altogether
df = df.dropna(subset=df.columns[-1])
df['mean'] = df[df.columns[7:]].mean(axis=1)
df['std'] = df[df.columns[7:]].std(axis=1)
df.head()
```

```python
from plotly.express import choropleth
for color in ['Continent', 'Hemisphere', 'Human Development Groups', 'UNDP Developing Regions', 'HDI Rank (2021)', 'mean', 'std']:
    choropleth(data_frame=df, locations='ISO3', color=color).show()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='mean', y='std', color='Human Development Groups', log_y=True, hover_name='Country')
```

The mean footprint tells us the level and the standard deviation tells us the volatility; mean consumption does a pretty good job of tracking development.
