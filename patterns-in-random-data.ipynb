```python
import warnings
import pandas as pd

DATA = '/kaggle/input/lottery-powerball-winning-numbers/Lottery_Powerball_Winning_Numbers__Beginning_2010.csv'

warnings.filterwarnings(action='ignore', category=FutureWarning)
# let's break out the ball numbers
df = pd.read_csv(filepath_or_buffer=DATA, parse_dates=['Draw Date'])
df['year'] = df['Draw Date'].dt.year
for index in range(6):
    df['n' + str(index)] = df['Winning Numbers'].str.split().str[index].astype(int)
df.head()
```

```python
len(df)
```

```python
from plotly import express

express.scatter(data_frame=df, x='Draw Date', y='Multiplier')
```

We do not have a full series, obviously.

We expect the ball numbers to look not entirely random, as the first five balls are reported in sorted order; we do expect the last ball (the power ball) to look random with a uniform distribution.

```python
for x in ['n0', 'n1', 'n2', 'n3', 'n4', 'n5']:
    express.histogram(data_frame=df, x=x, color='year').show()
```

The distribution of the power ball number does not look random; it looks like the power ball used to have numbers 1 through 39 until some point in 2015 and 1 through 26 after that.

```python
express.scatter(data_frame=df, x='n0', y='n1', color='year')
```

Our triangular scatter here is due to the fact that the first ball has the smallest number. It must have been a very strange night when that first ball was 52 and the second was 58.
