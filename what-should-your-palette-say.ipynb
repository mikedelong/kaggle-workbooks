```python
import pandas as pd

DATA = '/kaggle/input/antibiotic-use-in-livestock/countries-data-antibiotic-livestock.csv'
df = pd.read_csv(filepath_or_buffer=DATA)
df.head()
```

```python
df.shape
```

We have exactly eight rows; what if anything can we do with them?

```python
from seaborn import scatterplot

scatterplot(data=df, x='Year', y='Number of countries contributing data', hue='Number of countries contributing data', palette='viridis')
```

With only eight datapoints we're fine using the default shape and size for our scatter plot. And using the number of countries to determine the hue makes sense, because we're more interested in the progression and retreat of the phenomenon than we are in the passage of time.

```python
scatterplot(data=df, x='Year', y='Number of countries contributing data', hue='Year', palette='viridis')
```

The passage of time is interesting too, but a graph that colors by the year is less informative.
