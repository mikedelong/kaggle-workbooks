```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/visa-application-for-peoples-the-owlrd/visa_number_in_japan.csv')
df.head()
```

```python
df.info()
```

```python
from plotly.express import scatter
scatter(data_frame=df['Year'].value_counts(dropna=False).to_frame().reset_index(), x='Year', y='count')
```

Let's pick the most recent year and look at the data available.

```python
y2017_df = df[df['Year'] == 2017].copy().drop(columns=['Year'])
y2017_df = y2017_df[~y2017_df['Country'].isin({'total', 'others'})] # we just want countries
y2017_df.head()
```

```python
from plotly.express import scatter
scatter(data_frame=y2017_df, x='Number of issued', y='Number of issued_numerical', hover_name='Country', log_x=True, log_y=True)
# we have a big outlier so we need to use a log-log plot to be able to see anything.
```

Why are number issued and number issued (numerical) different? Who knows? But if we plot them against each other we can scatter countries in two dimensions without having to try to adjust explicitly for population of the source country, which is kind of nice, while sort of doing so implicitly.

```python
from plotly.express import scatter
scatter(data_frame=y2017_df, x='Number of issued', y='Number of issued_numerical', hover_name='Country', color='Ordinary',
        log_x=True, log_y=True, color_continuous_scale='bluered')
# we have a big outlier so we need to use a log-log plot to be able to see anything.
```

Ordinary visas are basically bimodal, with China and Vietnam in one mode and the rest of the world in another.

We can the same analysis and see how Japan manages its visas so that with the possible exception of China it concentrates visas of a particular type in a small handful of countries for different types.

```python
columns = [column for column in df.columns[5:].tolist() if column not in [key for key, value in y2017_df.isna().sum().to_dict().items() if value > 199]]
for column in columns:
    scatter(data_frame=y2017_df, x='Number of issued', y='Number of issued_numerical', hover_name='Country', color=column,
        log_x=True, log_y=True, color_continuous_scale='bluered', title=column).show()
```
