```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/100-million-random-values/100 Million Random Values.csv')
df.head()
```

```python
df.info()
```

```python
from plotly.express import imshow
from random import sample
from random import seed
seed(2023)
columns = sample(population=df.columns.tolist(), k=10)
imshow(img=df[columns].sample(n=10, random_state=2023).corr())
```

```python
from plotly.express import scatter_matrix
scatter_matrix(data_frame=df[sample(population=df.columns.tolist(), k=10)].sample(n=100, random_state=2023))
```

```python
from plotly.express import histogram
for column in sample(population=df.columns.tolist(), k=5):
    histogram(data_frame=df, x=column).show()
```

Oddly enough the values at the end of the range appear about half as often as they would if these numbers were uniform.
