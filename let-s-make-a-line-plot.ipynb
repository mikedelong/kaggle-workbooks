```python
import pandas as pd
import numpy as np
df = pd.read_csv(filepath_or_buffer='/kaggle/input/u-s-fed-funds-target-rate/US Fed Rate.csv', parse_dates=['Release Date'])
for column in ['Actual', 'Forecast', 'Previous']:
    df[column.lower()] = df[column].apply(func=lambda x: float(x.replace('%', '')) if x != '-' else np.nan)
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Release Date', y=['actual', 'forecast', 'previous'])
```
