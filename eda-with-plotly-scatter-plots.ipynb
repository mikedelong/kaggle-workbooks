```python
import pandas as pd
import math
MELBOURNE = '/kaggle/input/top-500-melbourne-eateries-tripadvisors-best/melbourne_ta_reviews.csv'
df = pd.read_csv(filepath_or_buffer=MELBOURNE).drop(columns=['category'])
df['log review count'] = df['numberOfReviews'].apply(func=math.log)
df.head()
```

```python
df['priceLevel'].value_counts()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='priceLevel', ).show()
```

```python
from plotly.express import scatter
scatter(y=df.sort_values(by='numberOfReviews')['numberOfReviews'],)
```

Most of our restaurants have fewer than 1000 reviews but our most-reviewed restaurants have over 6000; to get an interesting color according to the review count we need to use the log.

```python
scatter(data_frame=df, x='rankingPosition', y='rating', color='log review count', hover_name='name', hover_data=['address', 'priceLevel'])
```

Not surprisingly the 5/5 reviewed restaurants have far fewer reviews while the 3.5/5 restaurants have more.

```python
scatter(data_frame=df, y='rankingPosition', x='log review count', hover_name='name', hover_data=['address', 'priceLevel'], color='rating', trendline='ols')
```

Are we surprised that ranking (low is better) and review count are negatively correlated? Maybe. The R2 of our trendline is small, so we would say they are weakly correlated.
