```python
import pandas as pd
filename = '/kaggle/input/river-water-parameters/River water parameters.xlsx - Base de Datos.csv'
df = pd.read_csv(filepath_or_buffer=filename, parse_dates=['Date (DD/MM/YYYY)'])
df.columns = [' '.join(item.split()) for item in df.columns]
# we only have a few NaNs, and we want to visualize our data, so let's drop the NaNs
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='Sampling point').show()
histogram(data_frame=df, x='Hardness classification').show()
```

```python
from plotly.express import scatter
from umap import UMAP

model = UMAP(n_components=2, random_state=2023, verbose=True, n_jobs=1)
df[['u0', 'u1']] = model.fit_transform(X=df.drop(columns=['Date (DD/MM/YYYY)', 'Time (24 hrs XX:XX)', 'Sampling point', 'Hardness classification', 'Total Cl- (mg Cl-/L)']))
for color in ['Sampling point', 'Hardness classification', 'Total Cl- (mg Cl-/L)']:
    scatter(data_frame=df, x='u0', y='u1', color=color).show()
```
