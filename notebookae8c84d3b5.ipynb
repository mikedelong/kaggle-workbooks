```python
import pandas as pd

encoding = 'windows-1250'
CENTRAL = '/kaggle/input/global-debt-data/central_government_debt.csv'

central_df = pd.read_csv(filepath_or_buffer=CENTRAL, encoding=encoding, ).drop(columns=['indicator_name'])
central_t_df = central_df.T
central_t_df.columns = central_t_df.iloc[0]
central_t_df = central_t_df.tail(len(central_t_df)-1).reset_index()
central_t_df.head()
```

```python
from plotly.express import line
line(data_frame=central_t_df, x='index', y=central_t_df.columns[1:], log_y=False, height=1200)
```
