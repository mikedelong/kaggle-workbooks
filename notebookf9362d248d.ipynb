```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/luxury-watch-listings/Watches.csv', index_col=[0], low_memory=False)
df = df[df['price'] != 'Price on request']
df['currency'] = df['price'].str.get(0)
df['price_amount'] = df['price'].apply(lambda x: x if isinstance(x, float) else int(x[1:].replace(',', '')))
df.head()
```

```python
from plotly.express import histogram
for brand in ['Rolex', 'Omega', 'Seiko', 'Breitling', 'Cartier', 'Longines', 'TAG Heuer', 'Hublot', 'Audemars Piguet', 'Patek Philippe']:
    histogram(df[df['brand'] == brand], x='price_amount').show()
```
