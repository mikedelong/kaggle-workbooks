```python
import pandas as pd
from warnings import filterwarnings
filterwarnings(action='ignore', category=FutureWarning,)

AMZN = '/kaggle/input/2024-best-selling-sunglasses-on-amazon/best_sellers_amazon_2024_sunglasses.csv'
df = pd.read_csv(filepath_or_buffer=AMZN)
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='reviewsCount', log_y=True).show()
histogram(data_frame=df, x='stars').show()
histogram(data_frame=df.sort_values(by='stars'), x='price/value', color='stars').show()
```

Do we expect the number of stars to be a function of price? Maybe.

```python
from plotly.express import strip
strip(data_frame=df, x='price/value', y='stars', log_x=True, hover_name='title',  hover_data=['reviewsCount']).show()
strip(data_frame=df, x='price/value', y='stars', log_x=False, hover_name='title',   hover_data=['reviewsCount']).show()
```

Either way we slice it we have outliers; and at every star level there is a range of prices. 

```python
from plotly.express import box
box(data_frame=df, y='price/value', x='stars')
```

This plus the second histogram above tell us pretty much the whole story: we have lots of options at 4.2-4.5 stars, and lots of price options near 15.

```python
from plotly.express import scatter
scatter(data_frame=df.dropna(subset=['price/value']), x='reviewsCount', log_x=True, y='stars', hover_name='title', color='price/value')
```

We don't have a lot of prices (about 40% of our dataset), so when we remove null prices we don't have much data left.

```python
from plotly.express import density_heatmap
density_heatmap(data_frame=df, x='reviewsCount', y='stars', )
```

And not surprisingly almost all of our products, even the highly-rated ones, get closer to zero than to 50k reviews.
