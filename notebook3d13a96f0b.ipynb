```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/microsoft-stock-price2000-2023/MSFT(2000-2023).csv', parse_dates=['Date'])
df.head()
```

```python
df.info()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Close', 'Adj Close'])
```

```python
line(data_frame=df, x='Date', y=['Volume'], log_y=True,)
```

```python
from plotly.express import scatter
scatter(data_frame=df, y='Volume', x='Adj Close', color=df['Date'].dt.year, log_y=True, height=800)
```

```python
scatter(data_frame=df, x='Date', y=['Volume'], log_y=True, trendline='ols')
```
