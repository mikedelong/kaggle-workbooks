Let's load up our data; it has a non-utf-8 format, and there are spaces in the column names, and the dates are in a format that requires us to give pandas a hint.

```python
import pandas as pd

BANKS = '/kaggle/input/failed-banks-dataset/banklist.csv'

df = pd.read_csv(filepath_or_buffer=BANKS, encoding='windows-1250',)
df.columns = [column.strip() for column in df.columns]
df['Closing Date'] = pd.to_datetime(arg=df['Closing Date'], dayfirst=True)
df['year'] = df['Closing Date'].dt.year

df.head()
```

```python
from plotly import express

express.histogram(data_frame=df, x='Fund', ).show()
```

Fund sizes are bimodal, which is interesting.

```python
express.histogram(data_frame=df, x='Closing Date').show()
```

Closing dates are clustered in the years following the global financial crisis, which is not surprising.

```python
express.histogram(data_frame=df, x='State')
```

It's interesting to note that our state counts do not look entirely like a population map.

```python
express.scatter(data_frame=df, x='Closing Date', y='Fund', size='Fund', hover_name='Bank Name', color='year')
```

Fund sizes have two modes and closing dates have only one because fund sizes roughly doubled during the global financial crisis and haven't returned to pre-crisis levels.

```python
express.strip(data_frame=df, x='Closing Date',  hover_name='Bank Name', hover_data=['Acquiring Institution', 'Fund'])
```

This data defies simple visualization, as it has time, space, and magnitude dimensions.

```python
express.scatter(data_frame=df, x='State', y='Closing Date', size='Fund', hover_name='Bank Name', height=800, marginal_y='rug', color='Fund')
```
