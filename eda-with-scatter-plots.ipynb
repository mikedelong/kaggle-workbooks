```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/us-mail-statistics/Week 22 - US Mail - 1790 to 2017.csv', index_col=['index'])
df['income/office'] = df['Income'] / df['Number of Post Offices']
df['cost/office'] = df['Expenses'] / df['Number of Post Offices']
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Year', y=df.columns[3:])
```

```python
line(data_frame=df, x='Year', y='Pieces of Mail Handled')
```

```python
line(data_frame=df, x='Year', y='Number of Post Offices')
```

This is a fascinating curve; who would have guessed the number of post offices in 2016 was comparable to the number of post offices in 1865?

```python
from plotly.express import scatter
scatter(data_frame=df, x='Expenses', y='Income', color='Year', log_x=True, log_y=True)
```

```python
scatter(data_frame=df[df['Pieces of Mail Handled'].notnull()], x='Income', y='Pieces of Mail Handled', color='Year', log_x=True, log_y=True, trendline='ols')
```

```python
scatter(data_frame=df, x='Number of Post Offices', y='Income', color="Year", log_y=True)
```

```python
line(data_frame=df, x='Year', y=['income/office', 'cost/office'], log_y=True)
```

The money per office is vaguely sigmoid, with two flat regions and a growth region in the middle.
