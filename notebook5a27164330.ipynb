```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/30000-spotify-songs/spotify_songs.csv', index_col=['track_id'], )
df.head()
```

```python
df.info()
```

```python
from plotly.express import histogram
histogram(data_frame=df[df['track_popularity'] > 0], x='track_popularity', color='playlist_genre')
```

```python
df.columns
```

```python
for column in ['danceability',
       'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']:
    histogram(data_frame=df, x=column, color='playlist_genre').show()
```

```python
from plotly.express import bar
bar(data_frame=df['playlist_genre'].value_counts().to_frame().reset_index(), x='playlist_genre', y='count')
```

```python
from plotly.express import scatter
scatter(data_frame=df.sample(frac=1), x='loudness', y='energy', size='duration_ms', hover_name='track_artist', color='playlist_genre',
        facet_col_wrap=3, height=700, facet_col='key')
```

```python
from plotly.express import strip
strip(data_frame=df, x='playlist_genre', y='key')
```
