```python
import pandas as pd

NETFLIX = '/kaggle/input/netflix-movies-and-tv-shows/netflix_titles.csv'
USECOLS = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description']
df = pd.read_csv(filepath_or_buffer=NETFLIX, encoding='iso8859_10', usecols=USECOLS, index_col=['show_id'], )

# do a tiny bit of cleanup
df = df[~df['rating'].isin({'74 min', '84 min', '66 min'})]

df.head()
```

```python
from plotly import express

for column in ['type', 'release_year', 'rating', ]:
    express.histogram(data_frame=df, x=column).show()
```

```python
import warnings

warnings.filterwarnings(action='ignore', category=FutureWarning)
express.strip(data_frame=df, y='release_year', x='rating', color='type', hover_name='title', height=800)
```
