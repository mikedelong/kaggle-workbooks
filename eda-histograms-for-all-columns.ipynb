```python
import pandas as pd

AIR = '/kaggle/input/airlines-dataset/cleaned_air.csv'

df = pd.read_csv(filepath_or_buffer=AIR)

df.head()
```

```python
df.nunique()
```

```python
df.columns
```

```python
from plotly import express

for column in ['Operating Airline', 'Operating Airline IATA Code', 'Published Airline','Published Airline IATA Code',]:
    express.histogram(data_frame=df[column].value_counts().head(n=60).to_frame().reset_index(), x=column, y='count').show()
    express.histogram(data_frame=df[column].value_counts().tail(n=df[column].nunique() - 60).to_frame().reset_index(), x=column, y='count').show()
    
```

```python
from plotly import express

for column in ['GEO Summary', 'GEO Region', 'Landing Aircraft Type', 'Aircraft Body Type','Aircraft Manufacturer',]:
    express.histogram(data_frame=df, x=column).show()
```

```python
from plotly import express

for column in ['Landing Count', 'Total Landed Weight']:
    express.histogram(data_frame=df, x=column, log_y=True).show()
```
