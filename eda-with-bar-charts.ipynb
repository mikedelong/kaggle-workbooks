```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/decoding-the-stock-picks-of-american-lawmakers/US_CONGRESS_DATA.csv',
                parse_dates=['disclosure_date', 'transaction_date']).drop(columns=['ptr_link'])
df.head()
```

```python
df.info()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='disclosure_year', nbins=4)
```

```python
from plotly.express import bar
for column in [ 'owner', 'ticker', 'asset_description', 'type', 'amount', 'representative', 'district', 'state', 'industry', 'sector', 'party']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index().head(n=50), x=column, y='count').show()
```
