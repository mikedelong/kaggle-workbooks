```python
import pandas as pd
DATA = '/kaggle/input/color-distribution-of-sour-patch-kids/data.csv'
df = pd.read_csv(filepath_or_buffer=DATA, parse_dates=['date'])
df.head()
```

We have a couple of challenges here: one is to get the colors right; the other is to represent the fact that this is daily discontinuous data correctly. We need to use a color map and we need to get the shape of the histogram right.

```python
from plotly.express import histogram
histogram(data_frame=df, x='date', y=df.columns[1:-1], color_discrete_map={color: color for color in df.columns[1:-1]},
         nbins=(df['date'].max() - df['date'].min()).days)
```

The colormap is straightforward, but using daily bins gives us these tall thin poles of data.

```python
color_discrete_map={color: color for color in df.columns[1:-1]}
histogram(data_frame=df, x='date', y=df.columns[1:-1], color_discrete_map=color_discrete_map,)
```

Monthly bins give us a less accurate but more attactive view of the data.

```python
from plotly.express import scatter
scatter(data_frame=df, x='date', y=df.columns[1:-1], color_discrete_map=color_discrete_map, )
```

This is neat too; the y-axis shows the magnitude, but there is no way to make the size show the magnitude too.
