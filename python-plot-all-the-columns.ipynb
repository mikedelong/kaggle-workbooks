```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/pubg-stats-dataset/Pubg_Stats.csv', index_col=[0])
df.info()
```

```python
df.head()
```

```python
from plotly.express import scatter
for column in [item for item in df.columns if item not in {'Matches_Played', 'Rank', 'Player_Name', 'Weapons_Used'}]:
    scatter(data_frame=df, x='Matches_Played', y=column, color='Rank', hover_name='Player_Name', size='Weapons_Used',
           hover_data=df.columns).show()
```
