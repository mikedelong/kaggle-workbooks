```python
import pandas as pd

ELO = '/kaggle/input/nfl-elo-ratings/nfl_elo.csv'
df = pd.read_csv(filepath_or_buffer=ELO, parse_dates=['date'])
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
from seaborn import scatterplot

FIGSIZE = (14, 10)
PALETTE='viridis'
plt.figure(figsize=FIGSIZE)
scatterplot(palette=PALETTE, data=df, x='date', y='elo1_post', hue='team1', legend=False)
```

We really have too much data to make a single interesting picture here. Let's focus on a single season.

```python
from seaborn import lineplot
from warnings import filterwarnings

filterwarnings(action='ignore', category=FutureWarning)
plt.figure(figsize=FIGSIZE)
lineplot(palette=PALETTE, data=df[(df['season'] == 2019) ], x='date', y='elo1_post', hue='team1', legend=True)
```
