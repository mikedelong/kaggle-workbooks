```python
import pandas as pd

CARS = '/kaggle/input/car-prices-based-on-horsepower/mobil_mesin_harga.csv'

df = pd.read_csv(filepath_or_buffer=CARS)

df.head()
```

```python
df.info()
```

We don't have much data so let's make a small scatter plot.

```python
from plotly import express

express.scatter(data_frame=df, x='KekuatanMesin', y='Harga', trendline='lowess')
```

We have a trendline that looks rather linear; if we add a linear trendline instead how much of the data does it explain?

```python
from plotly import express

express.scatter(data_frame=df, x='KekuatanMesin', y='Harga', trendline='ols')
```

A linear trendline has an R2 of about 0.66, which explains why our data looks somewhat but not very linear. On the long end of our trendline the data is quite sparse.
