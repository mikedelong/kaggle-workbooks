```python
import pandas as pd

UBER = '/kaggle/input/uber-stock-data/UBER.csv'

df = pd.read_csv(filepath_or_buffer=UBER, parse_dates=['Date']).drop(columns=['Open', 'High', 'Low', 'Close'])
df['year'] = df['Date'].dt.year
df['dollars'] = df['Adj Close'] * df['Volume']
df.head()
```

Let's plot the price data first. Since we don't have a lot of data and the price hasn't increased much over the period of interest we don't need a log plot.

```python
import matplotlib.pyplot as plt
from seaborn import scatterplot

plt.figure(figsize=(10, 5))
scatterplot(data=df, x='Date', y='Adj Close', hue='year', palette='viridis')
```

Let's also plot the volume.

```python
plt.figure(figsize=(10, 5))
scatterplot(data=df, x='Date', y='Volume', hue='year', palette='viridis')
```

We have some big outliers, but we can see more detail if we use a log plot in the volume direction.

```python
plt.figure(figsize=(10, 5))
plt.yscale(value='log')
scatterplot(data=df, x='Date', y='Volume', hue='year', palette='viridis')
```

We have a synthetic dollar amount traded quantity we got by multiplying the price by the volume. What does that look like?

```python
plt.figure(figsize=(10, 5))
plt.yscale(value='log')
scatterplot(data=df, x='Date', y='dollars', hue='year', palette='viridis')
```

Because the price has mostly traded within a range the dollars look a lot like the volume.
