```python
from numpy import log10
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/top1000youtubers/youtubers_df.csv', index_col=['Rank'])
df['log10_visits'] = df['Visits'].apply(log10)
df['likes/visit'] = df['Likes'] / df['Visits']
df.head()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Visits', y='Likes', color='Country', hover_name='Username', log_x=True, log_y=True, trendline='ols',
       trendline_scope='overall')
```

Because the top channels are such outliers we need to use a log-log scale and that makes our OLS trendline look weird, but our data is surprisingly tightly clustered: the r2 for our OLS is 0.9.

```python
scatter(data_frame=df, x='Visits', y='Comments', color='Country', hover_name='Username', log_x=True, log_y=True, trendline='ols',
       trendline_scope='overall')
```

It always seems like viewer comments are more unpredictable than viewer likes, so there's more spread and the r2 from our OLS is much smaller.

```python
from plotly.express import bar
bar(data_frame=df.groupby(by='Country').size().reset_index().rename(columns={0:'count'}).sort_values(ascending=False, by='count').head(n=20),
x='Country', y='count', title='Top Twenty Countries')
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='log10_visits', color='Country',)
```

The log of total views looks vaguely Gaussian which is not surprising.
