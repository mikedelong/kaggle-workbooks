```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/airline-delay/Airline_Delay_Cause.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.manifold import TSNE
from plotly.express import scatter

airport_df = df.drop(columns=['year', 'month', 'carrier', 'carrier_name', 'airport_name']).dropna().groupby(by='airport').mean().reset_index()
airport_tsne = TSNE(n_components=2, verbose=1, random_state=2023)
airport_df[['t0', 't1']] = airport_tsne.fit_transform(X=airport_df.drop(columns=['airport']))
scatter(data_frame=airport_df, x='t0', y='t1', hover_name='airport', color='arr_flights')
```
