```python
import pandas as pd

CHEESE = '/kaggle/input/cheese-across-the-world/cheeses.csv'

df = pd.read_csv(filepath_or_buffer=CHEESE).drop(columns=['url'])
df['calcium (mg/100g)'] = df['calcium_content'].apply(func=lambda x: x if isinstance(x, float) else int(x.split()[0])) 

df.head()
```

Only some of our data is fit for purpose, and we have a lot of nulls; for some of our features we have too many different values to make good charts. Let's make a few bar charts with the features with a manageable number of values.

```python
from plotly import express

for column in ['milk', 'family', 'fat_content', 'calcium (mg/100g)', 'rind', 'color', 'vegetarian', 'vegan']:
    express.histogram(data_frame=df, x=column).show()
```

These are kind of boring, but we can make some scatter plots if we combine some of the categorical data with the calcium content.

```python
from plotly import express

for column in ['milk', 'family', 'rind']:
    express.strip(data_frame=df, x=column, y='calcium (mg/100g)', hover_name='cheese', log_y=True).show()
```
