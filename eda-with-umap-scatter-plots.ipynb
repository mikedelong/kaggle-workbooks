```python
import pandas as pd

CALLS = '/kaggle/input/google-fiber-internet-calls/Market_project_union.csv'

df = pd.read_csv(filepath_or_buffer=CALLS, parse_dates=['date_created'])
df.head()
```

```python
df.dropna().info()
```

```python
df.isna().sum()
```

We get more nulls as the number of contacts increases but we have nulls almost everywhere.

```python
from plotly import express

express.histogram(data_frame=df, x='date_created').show()
express.histogram(data_frame=df.dropna(), x='date_created').show()
```

```python
import warnings

warnings.filterwarnings(action='ignore', category=FutureWarning)
express.scatter(data_frame=df.dropna(), x='date_created', y=[column for column in df.columns if column.startswith('contacts')], log_y=True)
```

```python
import arrow
from plotly import express
from umap import UMAP

time_start = arrow.now()
reducer = UMAP(n_components=2, random_state=2024, transform_seed=2024, verbose=True, n_jobs=1, n_epochs=100)
umap_df = pd.DataFrame(data=reducer.fit_transform(X=df.dropna()[[column for column in df.columns if column.startswith('contact')]]), columns=['x', 'y'])
umap_df['contacts'] = df.dropna()['contacts_n'].tolist()
express.scatter(data_frame=umap_df, x='x', y='y', color='contacts').show()
print('UMAP done in {}'.format(arrow.now() - time_start))
```
