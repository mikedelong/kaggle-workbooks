```python
import pandas as pd
DATA = '/kaggle/input/uk-write-off-timeseries-data/UK Write-Off Data.csv'
df = pd.read_csv(filepath_or_buffer=DATA, parse_dates=['Date'], date_format='%d/%m/%Y')
df['year'] = df['Date'].dt.year
df.head()
```

```python
df.info()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=df.columns[1:], log_y=True).show()
```

We have data that varies substantially in magnitude; using a log scale shows all the data in a single graph, but the detail of the larger series is lost.

```python
line(data_frame=df, x='Date', y=df.columns[1], log_y=False).show()
line(data_frame=df, x='Date', y=df.columns[2:4], log_y=False).show()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='RPQB74D', y='RPQB75D', trendline='ols', hover_name='Date', color='year', log_x=False, log_y=False).show()
```

These two series look highly correlated in line plots, but in a scatter plot we see that in different periods of time they are in and out of phase.
