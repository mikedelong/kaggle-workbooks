```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/global-aviation-hub-dataset-of-airports-worldwide/airports .csv')
df.head()
```

```python
df.shape
```

```python
from plotly.express import bar
bar(data_frame=df.groupby(by='type').size().to_frame().rename(columns={0: 'count'}).sort_values(ascending=False, by='count'))
```

```python
from plotly.express import scatter_geo
scatter_geo(data_frame=df[df['type'] != 'closed'].sample(25000), lat='latitude_deg', lon='longitude_deg', color='type', hover_name='name')
```

For performance reasons we need to use a sample of the data as our scatterplot may become sluggish 

```python
scatter_geo(data_frame=df.sample(25000), lat='latitude_deg', lon='longitude_deg', color='elevation_ft', hover_name='name')
```
