```python
import pandas as pd

HMC = '/kaggle/input/honda-motor-stocks-prices/HMC.csv'
df = pd.read_csv(filepath_or_buffer=HMC, parse_dates=['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

```python
from plotly import express

express.line(data_frame=df, x='Date', y=['Open', 'High', 'Low', 'Close']).show()
express.line(data_frame=df, x='Date', y=['Close', 'Adj Close']).show()
express.line(data_frame=df, x='Date', y='Volume', log_y=True).show()
```

```python
express.scatter(data_frame=df, x='Date', y='Volume', trendline='lowess', log_y=True, color='year')
```

Our volume has outlier days both high and low, and the moving average trendline suggests we have a couple of different volume regimes.

```python
express.scatter(data_frame=df, x='Date', y='Adj Close', trendline='lowess', color='year')
```

Very broadly it looks like we have three different regimes: gains from March 2020 to August 2021, decline through about October 2022, and another run up through about March 2024. There are of course relatively short term reversals in each of these trends.

If we resample to monthly averages the result still has short-run losses in long-run gains and vice versa.

```python
express.scatter(data_frame=df[['Date', 'Adj Close']].set_index(keys=['Date']).resample('ME').mean().reset_index(), x='Date', y='Adj Close',
       trendline='lowess', )
```

```python
express.scatter(data_frame=df, x='Volume', y='Adj Close', color='year', log_x=True, log_y=False)
```

Because Honda has mostly risen over the course of the period of interest we see a very broad trend toward low values early and high values later.
