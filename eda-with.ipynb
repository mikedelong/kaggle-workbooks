```python
import pandas as pd

WEAR = '/kaggle/input/ethnic-wear-dataset/ethnicwear.json'

df = pd.read_json(path_or_buf=WEAR, ).set_index(keys=['product_id'])
df.head()
```

```python
from plotly import express

for column in ['brand', 'price', 'size', 'marketplace', 'discount', 'new_price']:
    express.histogram(data_frame=df, x=column).show()
```

```python
import warnings
warnings.filterwarnings(action='ignore', category=FutureWarning)

express.scatter(data_frame=df.sample(frac=1.0), x='price', y='new_price', log_x=True, log_y=True, color='brand', hover_name='product', height=800).show()
express.scatter(data_frame=df.sample(frac=1.0), x='price', y='new_price', log_x=True, log_y=True, color='size', hover_name='product', height=800).show()
express.scatter(data_frame=df.sample(frac=1.0), x='price', y='new_price', log_x=True, log_y=True, color='marketplace', hover_name='product', height=800).show()
```

The difference between the prices looks synthetic, and there doesn't seem to be any pattern in prices.
