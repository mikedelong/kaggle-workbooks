```python
import pandas as pd
BMRI = '/kaggle/input/bank-mandiri-stock-historical-price/BMRI.JK.csv'
df = pd.read_csv(filepath_or_buffer=BMRI, parse_dates=['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'])
```

Our data has one weird day in March 2023 that persists across all our prices series.

```python
line(data_frame=df, x='Date', y='Volume')
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Date', y='Volume')
```

```python
scatter(data_frame=df, y='Volume', x='Adj Close', color='year', log_x=True, log_y=True, trendline='ols')
```

Price x volume plots are always interesting; this one shows we have outliers in three directions, and that for most of the series the price (and volume) have been stuck in a trading range before apparently breaking out in 2023.

```python
scatter(data_frame=df, x='Date', y='Adj Close', trendline='lowess')
```

One weird trading day doesn't make much difference to our trendline which definitely shows a flat or downward direction early and an upward trend starting in mid-2021.
