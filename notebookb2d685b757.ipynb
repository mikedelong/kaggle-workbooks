```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/relative-search-volume-for-autismautism-spectrum/rsv_languages.csv', parse_dates=['date'])
df.head()
```

```python
df.info()
```

```python
from plotly.express import line
line(data_frame=df, x='date', y=df.columns[1:])
```

This is really noisy and we have a lot of data; let's see if resampling will smooth things out

```python
line(data_frame=df.set_index(keys=['date']).resample(rule='Q').mean().reset_index(), x='date', y=df.columns[1:], log_y=True, height=800)
```

Using a quarterly mean and a log plot in the Y direction smooths out the graphs and spreads them out, making them easier to tell apart.

```python
line(data_frame=df.set_index(keys=['date']).resample(rule='A').mean().reset_index(), x='date', y=df.columns[1:], log_y=True, height=800)
```

Using the annual mean probably goes too far.

```python
from plotly.express import scatter
scatter(data_frame=df.drop(columns=['date']).mean().to_frame().reset_index().merge(right=df.drop(columns=['date']).std().to_frame().reset_index(),
                                                                                   on='index', how='inner').rename(columns={'index': 'language', '0_x': 'mean', '0_y': 'stddev'}),
        x='mean', y='stddev', hover_name='language', color='mean')
```

A mean vs volatility plot is sometimes interesting for capturing essentially the same insight as we might find in the line plots above, rolling up the time series into a single point; the plot above tells us that German queries are consistently high, followed by English and (PRC) Chinese, while Turkish, Arabic, and Japanese queries are moderately high with high variability.
