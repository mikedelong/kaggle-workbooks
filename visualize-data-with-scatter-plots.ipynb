```python
import warnings
import pandas as pd

CRYPTO = '/kaggle/input/crypto-data/Crypto.csv'

warnings.filterwarnings(action='ignore', category=FutureWarning)

df = pd.read_csv(filepath_or_buffer=CRYPTO, parse_dates=['date_taken'])
df['price'] = df['price'].astype(float)
df.head()
```

```python
df.info()
```

Instead of time series data for a few currencies we have price data for a vast number of currencies. 

```python
from plotly import express

express.scatter(data_frame=df, x='date_taken', y='price', log_y=True, hover_name='abbr', )
```

Most cryptocurrencies have unit prices that are quite low.

```python
express.scatter(data_frame=df[df['price'] > 1], x='date_taken', y='price', log_y=True, hover_name='abbr', )
```

Cryptocurrencies with unit prices higher than a dollar are relatively rare. What percentage of the dataset are they?

```python
100 * len(df[df['price'] >= 1])/len(df)
```

Nine percent. Wow. How many stablecoins are in our dataset? We expect this number to be almost zero.

```python
len(df[df['price'] == 1])
```

How many are close?

```python
len(df[(df['price'] < 1.01) & (df['price'] > 0.99)])
```

```python
express.line(y=df.sort_values(by='price')['price'], log_y=True)
```

Almost none of our cryptocoins are worth more than a dollar.
