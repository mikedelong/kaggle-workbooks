```python
# most of the time we want to pay attention to every warning
# but right now 
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd

MOVIE = '/kaggle/input/52000-animation-movie-details-dataset-2024/Animation_Movies.csv'
df = pd.read_csv(filepath_or_buffer=MOVIE, parse_dates=['release_date'], index_col=['id'], thousands='.',
                ).drop(columns=['backdrop_path', 'homepage', 'poster_path'])

df = df[df['revenue'] != 0] # drop cases with no revenue
df = df[df['budget'] != 0] # drop cases with no budget 

df['money ratio'] = df['revenue'] / df['budget']
df['year'] = df['release_date'].dt.year
df.head()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='revenue', log_y=True).show()
histogram(data_frame=df, x='budget', log_y=True).show()
histogram(data_frame=df, x='money ratio', log_y=True).show()
histogram(data_frame=df, x='year')
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='budget', y='money ratio', log_x=True, log_y=True, color='year', hover_name='title', ).show()
scatter(data_frame=df, x='vote_count', y='popularity', log_x=True, log_y=True, color='year', hover_name='title', ).show()
scatter(data_frame=df, x='popularity', y='money ratio', log_x=True, log_y=True, color='year', hover_name='title', ).show()
scatter(data_frame=df, x='runtime', y='money ratio', log_x=True, log_y=True, color='year', hover_name='title').show()
```
