```python
import pandas as pd

JSON = '/kaggle/input/indian-stock-data-from-moneycontrol/stock_data_0.json'

df = pd.read_json(JSON)
df.head()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='THREATS', log_y=True)
```

Almost all of the companies in this dataset have no threats.

```python
from plotly.express import scatter
scatter(data_frame=df.dropna(subset='COST'), x='YEARLY_LOW', y='YEARLY_HIGH', color='COST', hover_name='NAME', log_x=True, log_y=True, height=900).show()
```

It looks like we have some missing data filled in with 10, which is an interesting choice. We also have some odd yearly low data, where the number 0.19 occurs more than we would expect.

```python
from plotly.express import histogram
for x in ['YEARLY_HIGH', 'YEARLY_LOW']:
    histogram(data_frame=df[df[x] < 1000], x=x, log_y=True).show()
    histogram(data_frame=df[df[x] < 11], x=x, log_y=True).show()
```

```python
from plotly.express import imshow
imshow(img=df[~df['COST'].isna()].corr(numeric_only=True))
```

```python
df['yearly ratio'] = df['YEARLY_HIGH'] - df['YEARLY_LOW']
histogram(data_frame=df[df['yearly ratio'] < 100], x='yearly ratio', log_y=True)
```

There is some strange data in the high/low ratios, but markets are strange places.
