```python
import pandas as pd

GOLD = '/kaggle/input/gold-stock-prices/goldstock.csv'

df = pd.read_csv(filepath_or_buffer=GOLD, index_col=[0], parse_dates=['Date'])
df['Daily Range'] = df['High'] - df['Low']
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Open', 'High', 'Low']).show()
line(data_frame=df, x='Date', y=['Volume',]).show()
line(data_frame=df, x='Date', y=['Daily Range',]).show()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Date', y='Open', trendline='lowess', color='Daily Range')
```

This is probably the nut graf, as it shows that it was possible to make money (in dollars) by buying gold before the start of the COVID pandemic and selling it at some point after the pandemic subsided. One way to interpret this is that the pandemic was inflationary.
