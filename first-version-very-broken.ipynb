```python
import pandas as pd

GALAXY = '/kaggle/input/sdss-galaxy-classification-dr18/sdss_100k_galaxy_form_burst.csv'
df = pd.read_csv(filepath_or_buffer=GALAXY, low_memory=False, header=1, index_col=['specobjid']).drop(columns=['objid', 'class'])
df.head()
```

```python
df.nunique()
```

```python
from umap import UMAP
from plotly import express

columns = [column for column in df.columns if column != 'subclass']

reducer = UMAP(n_components=2, random_state=2024, transform_seed=2024, verbose=True, n_jobs=1, n_epochs=100)
df[['x', 'y']] = pd.DataFrame(data=reducer.fit_transform(X=df[columns].fillna(value=0)))
express.scatter(data_frame=df, x='x', y='y', color='subclass', height=800, )
```
