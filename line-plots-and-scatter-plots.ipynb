```python
import pandas as pd
# we only have one series so we can drop the series name column
df = pd.read_csv(filepath_or_buffer='/kaggle/input/pakistan-trucks-sales/data_series.csv', parse_dates=['Observation Date']).drop(columns=['Series'])
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Observation Date', y='Observation Value')
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Observation Date', y='Observation Value', trendline='lowess', trendline_color_override='orange')
```

Because our data has inflection points a lowess trendline doesn't tell us much.

That July 2021 to December 2022 data looks odd, doesn't it?

```python
scatter(data_frame=df, x='Observation Date', y='Observation Value', trendline='rolling', trendline_options=dict(window=12),
        trendline_color_override='orange')
```

A moving average follows the data somewhat although it always trails the real data of course.

```python
scatter(data_frame=df.set_index('Observation Date').resample('Q').mean().reset_index(), x='Observation Date', y='Observation Value', 
        trendline='rolling', trendline_options=dict(window=4),
        trendline_color_override='orange')
```

A quarterly moving average smooths out a lot of the volatility while retaining the basic shape.
