```python
import pandas as pd
from shapely.geometry import Point

DATA = '/kaggle/input/doctors-locations-vienna/ARZTOGD.csv'
df = pd.read_csv(filepath_or_buffer=DATA).drop(columns=['FID','SE_ANNO_CAD_DATA'])
df.head()
```

```python
df.info()
```

```python
# https://gis.stackexchange.com/a/404301
from geopandas import GeoDataFrame
from shapely import wkt
df['geometry'] = df['SHAPE'].apply(wkt.loads)
gdf = GeoDataFrame(df, crs='epsg:4326')
gdf['latitude'] = gdf['geometry'].y
gdf['longitude'] = gdf['geometry'].x
```

```python
from plotly.express import scatter_geo
scatter_geo(data_frame=gdf, lat='latitude', lon='longitude', fitbounds='locations',basemap_visible=True,
           hover_name='NAME', hover_data=['ADRESSE', 'FACH', 'TELEFON', 'INTERNET'], color='OBJECTID')
```
