```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/united-nations-refugee-data-unhcr/United Nations Refugee Data.csv')
df.head()
```

```python
from plotly.express import histogram
histogram(data_frame=df, x='Year')
```

```python
from plotly.express import bar
for column in ['Country of origin (ISO)', 'Country of asylum (ISO)']:
    bar(data_frame=df[column].value_counts().to_frame().reset_index(), x=column, y='count').show()
```

```python
from plotly.express import parallel_categories
parallel_categories(data_frame=df[df['Refugees under UNHCR\'s mandate'] > 100000], dimensions=['Country of origin (ISO)', 'Country of asylum (ISO)'], color='Refugees under UNHCR\'s mandate',
                   height=1000)
```

```python
total_df = df[['Country of origin (ISO)', 'Country of asylum (ISO)', 'Refugees under UNHCR\'s mandate']].groupby(by=['Country of origin (ISO)', 'Country of asylum (ISO)', ]).sum().reset_index()
parallel_categories(data_frame=total_df[total_df['Refugees under UNHCR\'s mandate'] > 100000],
                   dimensions=['Country of origin (ISO)', 'Country of asylum (ISO)',], color='Refugees under UNHCR\'s mandate',
                   height=1000)
```

```python
from plotly.express import choropleth
for column in ['Country of origin (ISO)', 'Country of asylum (ISO)']:
    choropleth(data_frame=total_df, locations=column, color='Refugees under UNHCR\'s mandate').show()
```

```python
bar(data_frame=total_df.drop(columns=['Country of asylum (ISO)']).groupby(by='Country of origin (ISO)').sum().reset_index(), x='Country of origin (ISO)', y='Refugees under UNHCR\'s mandate',)
```
