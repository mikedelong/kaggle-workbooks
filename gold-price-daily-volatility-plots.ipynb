```python
import pandas as pd
GOLD = '/kaggle/input/gold-coin-prices/1.csv'
df = pd.read_csv(filepath_or_buffer=GOLD, parse_dates=['Date'])
df['Change'] = df['Close'] - df['Open']
df['rel change'] = df['Change'] / df['Open']
df['year'] = df['Date'].dt.year
df.head()
```

```python
df.info()
```

```python
from plotly.express import line
for log_y in [False, True]:
    line(data_frame=df, x='Date', y=['Open', 'High', 'Close'], log_y=log_y).show()
```

We see different levels of detail in the prices with and without the log plot.

```python
from plotly.express import scatter
scatter(data_frame=df, x='Date', y='Change', color='year').show()
scatter(data_frame=df, x='Date', y='rel change', color='year').show()
```

The biggest absolute price changes have happened recently at high price levels, but the largest relative price changes happened in 2018.

```python
from plotly.express import histogram
histogram(data_frame=df, x='rel change', log_y=True, marginal='box', )
```

The median daily change is zero; almost all of the daily changes are essentially zero. Single-day price changes of +23% or -14% are really wild.
