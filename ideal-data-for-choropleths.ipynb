```python
import pandas as pd

FILENAME = '/kaggle/input/world-population-statistics-2023/world_population_data.csv'
df = pd.read_csv(filepath_or_buffer=FILENAME, index_col=['rank'])
df['growth (pct)'] = df['growth rate'].apply(func=lambda x: float(x[:-1]))
df['world population (pct)'] = df['world percentage'].apply(func=lambda x: float(x[:-1]))
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
from plotly.express import choropleth
choropleth(data_frame=df, locations='cca3', color='continent')
```

This is a very complete dataset; most of the country-based data we see is missing at least a handful of countries.

```python
for color in ['area (km²)', 'density (km²)', 'growth (pct)', 'world population (pct)', '2023 population', '2022 population',
       '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population',]:
    choropleth(data_frame=df, locations='cca3', color=color).show()
```
