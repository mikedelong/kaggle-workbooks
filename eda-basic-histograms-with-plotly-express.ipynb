Let's make some histograms where our data is amenable. Some of our data is so clustered around its mode that we want to use a log plot in the Y/count direction.

```python
from plotly import express

IGNORE = {'AgeGyr'}
LOGY = {'PlanetaryMassJpt', 'PeriodDays', 'SemiMajorAxisAU', 'DiscoveryYear', 'HostStarRadiusSlrRad'}
NUMERIC = {'float64', 'int64'}

for column in [key for key, value in df.dtypes.to_dict().items() if str(value) in NUMERIC and key not in IGNORE]:
    express.histogram(data_frame=df, x=column, log_y=column in LOGY).show()
```
