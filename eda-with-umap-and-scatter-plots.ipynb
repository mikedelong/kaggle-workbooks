```python
import pandas as pd

SCM = '/kaggle/input/supply-chain-management/SCM Dataset.xlsx'

df = pd.read_excel(io=SCM, sheet_name='DATASET')
df['Cost of Goods Sold (COGS)'] = df['Cost of Goods Sold (COGS)'].apply(func=lambda x: float(x.replace('$', '').replace(',', '').replace('B', '')))
df.head()
```

```python
import arrow
import warnings
from plotly import express
from umap import UMAP

COLUMNS = [key  for key, value in df.dtypes.to_dict().items() if str(value) in {'float64', 'int64'} and df[key].isna().sum() == 0]
OBJECTS = [key  for key, value in df.dtypes.to_dict().items() if str(value) not in {'float64', 'int64'}]

time_start = arrow.now()
warnings.filterwarnings(action='ignore', category=FutureWarning)
reducer = UMAP(n_components=2, random_state=2024, transform_seed=2024, verbose=True, n_jobs=1, n_epochs=1000)
umap_df = pd.DataFrame(data=reducer.fit_transform(X=df[COLUMNS]), columns=['x', 'y'])
umap_df[OBJECTS] = df[OBJECTS].copy()
for color in OBJECTS[1:]:
    express.scatter(data_frame=umap_df, x='x', y='y', hover_name='Company Name', color=color).show()
print('UMAP done in {}'.format(arrow.now() - time_start))
```
