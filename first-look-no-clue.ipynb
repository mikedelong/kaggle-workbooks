```python
import pandas as pd

FAILS = '/kaggle/input/device-failure/device_failure.csv'
df = pd.read_csv(filepath_or_buffer=FAILS, parse_dates=['date'])
df['Failure'] = df['failure'].astype(bool)
df.head()
```

```python
import warnings
from plotly import express
warnings.filterwarnings(action='ignore', category=FutureWarning)

express.pie(data_frame=df, names='failure', color='failure')
```

Not surprisingly we are looking for a really rare phenomenon.

```python
df.columns
```

```python
from umap import UMAP
from plotly import express

columns = [column for column in df.columns if column.startswith('attribute')]
columns = [
    'attribute1', 
    'attribute2', 
#     'attribute3',
#        'attribute4', 'attribute5', 
    'attribute6', 
#     'attribute7', 'attribute8',
#        'attribute9'
]
target = 'Failure'

sample_df = df.sort_values(ascending=False, by='Failure').head(n=1000).copy()
reducer = UMAP(n_components=2, random_state=2024, transform_seed=2024, verbose=True, n_jobs=1, n_epochs=200)
sample_df[['x', 'y']] = pd.DataFrame(data=reducer.fit_transform(X=sample_df[columns]))
# express.scatter(data_frame=sample_df, x='attribute1', y='attribute6', color=target, hover_name='device').show()
express.scatter(data_frame=sample_df, x='x', y='y', color=target, ).show()
```
