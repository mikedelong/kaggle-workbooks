```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/daily-google-news/2023_9.csv', parse_dates=['DateTime'])
df['month'] = df['DateTime'].dt.month
df['year'] = df['DateTime'].dt.year
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
from plotly.express import bar
for column in ['year', 'month', 'Publisher', 'Category']:
    bar(data_frame=df[column].value_counts().nlargest(n=50).to_frame().reset_index(), x=column, y='count', 
        log_y=column in {'year', 'month'}).show()
```

```python
from plotly.express import scatter
publisher_df = df['Publisher'].value_counts().to_frame().reset_index().reset_index()
scatter(data_frame=publisher_df, x='index', y='count', hover_name='Publisher')
```

```python
scatter(data_frame=publisher_df[publisher_df['count'] > 100], x='index', y='count', hover_name='Publisher')
```

```python
from matplotlib.pyplot import subplots
from matplotlib.pyplot import axis
from matplotlib.pyplot import imshow
from wordcloud import WordCloud
subplots(figsize=(12, 12))
imshow(X=WordCloud(random_state=2023, height=1200, width=1200 ).generate(text=' '.join(df['Publisher'].values.tolist())), )
axis('off')
```

```python
subplots(figsize=(12, 12))
imshow(X=WordCloud(random_state=2023, height=1200, width=1200 ).generate(text=' '.join(df['Title'].values.tolist())), )
axis('off')
```
