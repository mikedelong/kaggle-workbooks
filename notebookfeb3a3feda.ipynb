```python
import pandas as pd

CHOCOLATE = '/kaggle/input/chocolate-ratings/chocolate.csv'
df = pd.read_csv(filepath_or_buffer=CHOCOLATE)
df['cocoa'] = df['cocoa_percent'].apply(func=lambda x: float(x.replace('%', ''))/100)
df.head()
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='cocoa', y='rating', color='review_date', )
```
