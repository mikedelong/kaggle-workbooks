```python
import pandas as pd

KAGGLE = '/kaggle/input/kaggle-top-datasets/kaggle-preprocessed.csv'

df = pd.read_csv(filepath_or_buffer=KAGGLE, index_col=[0], parse_dates=['Date']).drop(columns=['Author_id', 'Dataset_link', 'Time'])
df.head()
```

```python
from plotly import express

express.histogram(data_frame=df, x='Day', color='Medals')
```

Top datasets are more likely to be added early in the week.

```python
from plotly import express

express.histogram(data_frame=df, x='Type_of_file')
```

Almost all of our data is CSV and/or has at least one file that is CSV.

```python
express.histogram(data_frame=df, x='No_of_files', log_y=True)
```

Almost all of our datasets have only a handful of files, but the datasets with lots of files make that hard to see. Let's look at the distribution on the small end,

```python
express.histogram(data_frame=df[df['No_of_files'] < 30], x='No_of_files', log_y=True)
```

Almost all Kaggle datasets have ten files or less.

```python
import warnings

warnings.filterwarnings(action='ignore', category=FutureWarning)

express.histogram(data_frame=df, x='Upvotes', color='Medals', log_y=True)
```

We need to do something similar with upvotes/medals,

```python
express.histogram(data_frame=df[df['Upvotes'] < 100], x='Upvotes', color='Medals', log_y=True)
```

Oh that's interesting; there are datasets with more than twenty upvotes but no medals. This should not be happening, as only five upvotes are required to get a bronze medal, provided the accounts up-voting are of some minimum age. 

```python
df[(df['Upvotes'] > 20) & (df['Medals'] == 'No Medal')]['Author_name'].value_counts()
```

How many datasets have more than twenty upvotes but no medals?

```python
len(df[(df['Upvotes'] > 20) & (df['Medals'] == 'No Medal')]['Author_name'])
```

That seems like a lot. 

```python
express.histogram(data_frame=df, x='Medals') 
```

What would we expect sight unseen for this breakdown? Getting a silver medal for a dataset is hard and getting a gold medal is really hard; maybe the surprise here is that there are so many top datasets with no medals.

```python
express.histogram(data_frame=df, x='Usability', color='Medals', nbins=200)
```

We expect high-rated datasets to have high usability; the question might be why we have so many high-rated datasets with low usability.

```python
express.scatter(data_frame=df, x='Usability', y='Upvotes', color='Medals', log_y=True, hover_name='Dataset_name')
```

This is probably the nut graf because it shows so much data and it shows the loose relationship between upvotes and medals, and also usability and upvotes.
