```python
import pandas as pd

KAGGLE = '/kaggle/input/kaggle-top-datasets/kaggle-preprocessed.csv'

df = pd.read_csv(filepath_or_buffer=KAGGLE, index_col=[0], parse_dates=['Date'])
df.head()
```

```python
df.info()
```

```python
from plotly import express

express.histogram(data_frame=df, x='Type_of_file')
```

Almost all of our data is CSV and/or has at least one file that is CSV.

```python
express.histogram(data_frame=df, x='No_of_files', log_y=True)
```

Almost all of our datasets have only a handful of files, but the datasets with lots of files make that hard to see. Let's look at the distribution on the small end,

```python
express.histogram(data_frame=df[df['No_of_files'] < 30], x='No_of_files', log_y=True)
```

Almost all Kaggle datasets have ten files or less.

```python
import warnings

warnings.filterwarnings(action='ignore', category=FutureWarning)

express.histogram(data_frame=df, x='Upvotes', color='Medals', log_y=True)
```
